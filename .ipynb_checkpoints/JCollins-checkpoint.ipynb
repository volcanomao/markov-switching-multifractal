# Abstract

The Markov-switching multifractal stochastic volatility model (MSM) of Calvet & Fisher (2004, 2008a)) permits the parsimonious specification of a high-dimensional state space. I show that out-of-sample performance improved when the state space was expanded, high-frequency data used, and microstructure noise taken into account. I enabled maximum likelihood estimation and analytical forecasting with up to 13 volatility frequencies and over 8,000 states, some eight times higher than previous literature, by implementing (coding) the model in new ways. I introduced a stochastic algorithm that combined heuristic procedures with local searches to perform an enhanced exploration of the state space in conjunction with local optimization. Rigorous preparation and cleansing of data, sparse sampling, and return innovations weighted by the respective depth of the best bid and ask, mitigated microstructure noise. These developments resulted in a well-specified model, better able to use the increased information provided to it by large high frequency (HF) datasets. In-sample model selection tests showed statistically significant monotonic improvement in the model as more volatility components were introduced. MSM(13) was compared to the relative accuracy of out-of-sample forecasts produced by a realized volatility measure using the heterogeneous autoregressive (HAR) model of Corsi (2009). MSM(13) provided comparatively better, statistically significant, forecasts than HAR most of the time at 1-hour, 1-day, and 2-day horizons for equity HF (Apple and J.P.Morgan) and foreign exchange HF (EURUSD) returns series. MZ regressions showed little sign of bias in the MSM(13) forecasts. These results suggest MSM may provide a viable alternative to established realized volatility estimators in high-frequency settings.

# Introduction

Deterministic (G)ARCH-class models recognize that there may be important nonlinearity, asymmetry, and long memory properties of volatility and that returns can be non-normal with a variety of parametric and nonparametric distributions (see Engle (2002) and Bollerslev, Engle & Nelson (1994) for extensive surveys of the (G)ARCH model landscape). Closely related, but econometrically distinct, stochastic volatility (SV)-class models have a different data-generating process that makes them more convenient, for some purposes, but, in some circumstances, more difficult to estimate (see for example Harvey, Ruiz, & Shephard (1994) and Taylor (1994)). In a linear framework, both classes may be considered as different representations of the same process; but in non-linear settings, the alternative specifications are not equivalent, although they can be close approximations. Partly this is due to three general characteristics of financial returns: unpredictability; surprisingly large numbers of extreme values (fat tails); and extreme and quiet periods clustering in time (volatility clustering). These are the characteristics for which (G)ARCH and SV models were designed, so it is perhaps not surprising that these models have empirically proven over nearly four decades to faithfully represent these dynamics. 
$\\[0.1in]$

The multifractal approach is a compelling alternative for modeling financial returns, and volatility in particular. Multifractal analysis reveals that returns observed at different time horizons exhibit several forms of scale invariance, based on multifractal moment-scaling. Calvet & Fisher (2002), Mandelbrot, Calvet & Fisher (1997a, 1997b, 1997c), and Ghashgaie, Breymann, Peinke, Talkner, & Dodge (1996) found early evidence that the moments of the absolute value of price changes, $\mathbb{E} \left( {\mid p \left( t + \Delta t \right) - p \left( t \right) \mid}^q \right)$, scale as power functions of the horizon $\Delta t$. This observation motivated the construction of early multifractal diffusions (see for example Bacry, Delour & Muzy (2001) and Calvet & Fisher (2001)). Subsequently, the Markov-Switching Multifractal, $\mathrm{MSM} \left( \bar k \right)$ or MSM,  of Calvet & Fisher (2004, 2008a) was shown to capture well not only the dynamics of returns series - unpredictability, fat tails, long memory - but also moment scaling (see for example Calvet & Fisher (2008a, 2004, 2001), and Liu, Matteo & Lux (2007)). MSM assumes that the size of a price change is driven by a product of components that have invariant distributions but different degrees of persistence, where each component follows a Markov chain. MSM thus constructs a multifractal measure stochastically over time.
$\\[0.1in]$

MSM permits the parsimonious specification of a high-dimensional state space. It accommodates an arbitrarily large number of volatility components, or frequencies, $\bar k$, whilst retaining a small and constant number of parameters; in its binomial form, requiring just four. When volatility components have a discrete distribution, MSM is a latent Markov chain and the parameters can be estimated by maximizing the closed-form likelihood of the returns series. But this can make MSM computationally expensive. MSM's doubly stochastic transition matrix, $A$, contains $2^{\bar k} \ast 2^{\bar k}$ elements and grows quickly with $\bar k$. For example, with 13 frequencies, the transition matrix has cardinality $2^{26} \approx$ 67 million elements. High $\bar k$ settings therefore present a formidable computational challenge. 
$\\[0.1in]$

To circumvent this, researchers have confined the use of maximum likelihood estimation to a moderate number of volatility components (for example, Alausa (2014) studies $\bar k \leq 8$, Hainault (2012) $\bar k \leq 7$, Idier (2009, 2008) $\bar k \leq 8$, and Calvet & Fisher (2004) $\bar k \leq 10$). Alternative computational methods to deal with high-dimensional state spaces were introduced, but involve some sacrifice of performance. Calvet, Fisher & Thomson (2006) used a particle filter for inference and forecasting using simulations, following literature on Markov chains (for example Chib, Nardari & Shephard (2002), Pitt & Shephard (1999), and Jacquier, Polson & Rossi (1994)). Lux (2008) developed a GMM estimator, thus enabling less computationally demanding linear forecasting. With this setup Lux studied $\bar k \leq 20$, demonstrating improved forecasting accuracy up to $\bar k = 15$; but these results were obtained in a restricted setting, with two parameters arbitrarily fixed. Tractability was similarly addressed by use of GMM estimation and best linear forecasts by Segnon, Lux & Gupta (2015), Lux, Morales-Arias & Sattarhoff (2011), Lux & Morales-Arias (2009), Liu, Matteo & Lux (2007), and Lux & Kaizoji (2007)). Bacry, Kozhemyak & Muzy (2008) also used a GMM estimator. Calvet and Fisher (2002) showed that efficiency gains from ML estimation are substantial relative to a variety of simulated method of moments estimators, with improvement in estimation efficiency confirmed by Lux (2008). In this study I maintain Bayesian updating and a closed-form likelihood function up to 13 components. One of the contributions to the multifractal literature that I make is to show that this approach, previously considered infeasible, can be viable and lead to improved out-of-sample performance when used in conjunction with high frequency (HF) data that accounts for microstructure noise. 
$\\[0.1in]$

To mitigate microstructure noise, I rigorously prepared, cleansed, and sparsely sampled HF data. Careful data cleansing is one of the most important aspects of volatility estimation with high frequency data (Barndorff-Nielsen, Hansen, Lunde & Shephard (2009) expand considerably upon this point) and it is not always straightforward to construct a tractable time series from raw high frequency data. The methods I employed to prepare the data were therefore not trivial. I followed the data cleansing procedures set out in Bollerslev, Hood, Huss & Pederson (2017), Barndorff-Nielsen, Hansen, Lunde & Shephard (2009) and Brownlees & Gallo (2006), which are in turn consistent in many respects with Dacorogna, Gencay, Muller, Olsen, & Pictet (2001). Variants of these procedures are also followed by several other researchers, including Bollerslev, Li & Xue (2018), Ait-Sahalia & Jacod (2012), Ait-Sahalia, Mykland & Zhang (2010), Barndorff-Nielsen, Hansen, Lunde & Shephard (2007, 2008), and Hanson & Lunde (2006). Using prepared data, I computed returns using a weighted mid-price (WMP), incorporating the respective depth of the best Bid and Ask. Volume at the best Bid and Ask has been shown to be a strong predictor of the next traded price. This feature of imbalances has been reported by Stoikov (2020), Dacorogna (2018), Lehalle & Mounjid (2018), Stoikov & Waeber (2016), Gould & Bonart (2015), Lehalle & Laruelle (2014), Lipton, Pesavento & Sotiropoulos (2013), Burlakov, Kamal & Salvadore (2012), Robert & Rosenbaum (2012), Avellaneda, Reed & Stoikov (2011), and Gatheral & Oomen (2010)). However, my primary justification for use of WMP is empirical, not theoretical. WMP has been widely used in practice (see for example Stoikov (2020), Dacorogna (2018), Lehalle & Mounjid (2018), and Gatheral & Oomen (2010)). It has also been used by other researchers (for example, Ait-Sahalia, Mykland & Zhang (2010)).
$\\[0.1in]$

High-dimensional state spaces necessitate high performance computing. I therefore implemented some MSM functions on a graphics processing unit (GPU), using approximately 2,000 cores or parallel processors. I further optimized and accelerated the model's operation on the GPU using high performance array libraries (for example, CuPy). I implemented Numba, a just-in-time (JIT) compiler, to translate some of my Python functions to optimized machine code at runtime, in turn achieved via use of LLVM compiler libraries. I also used memoization, a technique for recording intermediate results to avoid repeated calculations. I introduced a two-step basin-hopping method, following Wales & Doye (1997), into the estimation process. The basin-hopping framework has been broadly used outside of finance for hard nonlinear optimization problems with multiple variables and modalities. It uses a stochastic algorithm, combining heuristic procedures with local searches. This performs an enhanced exploration of the MSM state space, conducted as a series of perturbations, followed by local optimization. The algorithm is iterative, with each cycle comprising a random perturbation of the binomial MSM parameter vector $\theta = \left( b, \: m_0, \: \: \gamma_{\bar k}, \: \bar \sigma \right)$, local minimization, and acceptance or rejection of the new parameter vector based upon the minimized function value. These steps resulted in a well specified model and more robust estimation. Compute time for the benchmark JPYUSD daily returns dataset (this benchmark was chosen in order to replicate Calvet & Fisher (2004)) with $\bar k = 4$ was reduced from 00:08:51.624 [hh:mm:ss] (nearly 9 minutes) to 00:00:00.927 [hh:mm:ss] (just under 1 second).
$\\[0.1in]$

I compared out-of-sample forecasts from the MSM to the relative accuracy of out-of-sample forecasts produced by a realized volatility (RV) measure. I used the heterogeneous autoregressive (HAR) forecasting model of Corsi (2009) for the comparison. The HAR model is popular in practice because it captures long-memory-type properties of quadratic variation, while being simpler to estimate than fractionally integrated processes. The HAR model has been shown to consistently provide strong out-of-sample volatility forecasts (see for example Clements & Preve (2019), Bollerslev, Patton & Quaedvlieg (2018, 2016), and Liu, Patton & Sheppard (2015)). The multifractal literature to date does not appear to have directly evaluated the performance of a multifractal model with HF data and/or RV.  Where HF data has been used (for example Lux, Morales-Arias & Sattarhoff (2011)), it is in the context of incremental model extension within a more widely used setting, or in settings where MSM is used to study inter-trade durations (see Zikes, Barunik & Shenai (2018), Aldrich, Heckenbach & Laughlin (2014), and Chen, Diebold & Schorfheide (2013)). 
$\\[0.1in]$

MSM's construction theoretically accommodates an infinity of frequencies. Calvet & Fisher (2004, 2008a) showed that MSM's likelihood function levels off when the number of components reaches the range $\bar k = 8, 9, 10$ (a result not contradicted by other researchers using less components: see for example Alausa (2014), Hainault (2012), Idier (2009, 2008)). According to Calvet & Fisher, improvement of the likelihood function was likely to be limited as $\bar k$ goes to infinity and the frequency growth rate parameter, $b$, correspondingly goes to unity. They posit that the martingale convergence theorem with appropriate tightness conditions implies convergence of the time sequence to a non-degenerate time deformation process. The limiting multifractal diffusion is thus driven by countably many components, and its sample paths contain a continuum of local Holder exponents in any time interval. My results don't necessarily contradict this, they simply suggest that the convergence may occur somewhat later. I observe that frequency growth rate $b$ falls, but slowly and monotonically through the range 1.5 to 1.2 for $\bar k = 11$ to $\bar k = 13$. The small incremental falls are sufficient to provide significant improvements in the likelihood because they are accompanied by a large expansion of the state space. Consider, when $\bar k$ was increased from 11 to 12, the number of volatility states doubled from 2048 to 4096, the state space quadrupled, so that matrix $A$ was constructed with 16,777,216 elements. When $\bar k$ was increased again to 13, MSM had 8,192 components, the state space again quadrupled, and $A$ expanded to nearly 70 million elements. This provided space to capture more of the patterns in the volatility of the complex HF time series and thus both estimation and performance improved in lockstep. At some point the benefit of state space expansion must satiate (note that, theoretially, when $\bar k = \infty$, the state space is a continuum). But the latent information in the volatility of HF time series may benefit from yet larger state space still, beyond the $\bar k = 13$ boundary of this study, to fully capture different tje volatility regimes that may be observed in an HF time-series - long memory, clustering, fat tails, outliers and extreme events, macro-economic cycles, seasonality, diurnal patterns, market events, and news announcements, to name but a few.
$\\[0.1in]$

My main insights and contributions to the multifractal literature may be summarized as follows. Firstly, I expanded the state space to allow more volatility components to be applied in likelihood-based estimation and forecasting. Secondly, I enabled this to be done with HF data. Thirdly, I specifically accounted for and mitigated microstructure noise. Fourthly, to enable the previous three contributions, I introduced several implementation-related developments to speed up the model. This all resulted in a well-specified model, better able to use the increased information provided to it by large amounts of HF data. The results are encouraging. In-sample model selection tests showed statistically significant monotonic improvement in the model as more volatility components were introduced. MSM(13) provided comparatively better, statistically significant, forecasts than HAR most of the time at 1-hour, 1-day, and 2-day horizons for equity HF (Apple and J.P.Morgan) and foreign exchange HF (EURUSD) returns series. MZ regressions showed little sign of bias in the MSM(13) forecasts.
$\\[0.1in]$

The rest of this paper is organized as follows. Sections 3 and 4 review the literatures on multifractality and microstructure noise respectively. Section 5 describes the preparation of the high frequency returns and describes them empirically. Section 6 develops the Multifractal volatility model in the presence of microstructure noise and fully defines the setting, out-of-sample performance evaluation, and relative performance assessment benchmarks. Section 7 provides empirical results for AAPL, JPM, EURUSD 1-minute returns and DEXJPUS daily returns and compares MSM out-of-sample performance to that of HAR. Section 8 concludes.

# Multifractality

Multifractal formalisms provide powerful metrics with which the econometrician may quantify the complex spatio-temporal fluctuations that occur in financial systems, asset returns in particular. Multifractality reflects the presence of both long-range dependence and hierarchical organization; fractals use invariance principles to parsimoniously model complex objects at multiple scales. A particularly important area of application is to model and explain long-memory processes that occur in the time domain, such as those found in financial time series (see, for example, Beran's (1994) and Mandelbrot's (1997) early expositions). Good models describe these multifractal scaling characteristics as well as their endogenous origin: That is, the multifractal properties result from the explicit mechanism that future values of the process are (nonlinearly) influenced by the whole past history. In this section I succinctly review the empirical literature on use of fractals in finance, which my work aims to extend. 
$\\[0.1in]$

Mandelbrot first discovered evidence of fractal behavior in financial returns, household income, and household wealth in the late 1950’s and early 1960’s. In finance, the distribution of price changes is of great interest because it determines both the risk and the potential gains from an investment. Different investors may measure price changes at different horizons. For example, HF traders measure prices in sub-second intervals, and I deal with some of the issues related to operations within these timeframes in some detail in this paper. On the other hand, long-term investors such as sovereign wealth funds and endowments consider investments over decades. But complex decisions are also required during the longer horizon.
$\\[0.1in]$

To illustrate this interplay between short and longer-term dynamics, Pal Haugerud, Director General of Asset Management at the Norwegian Ministry of Finance said, in 2012: "This fund [the Norwegian Sovereign Wealth Fund] has characteristics that enable us to accept volatility. It is very unlikely that the Norwegian state will need to withdraw large sums of money from the fund in the short-term. We do not depend on short-term financing, and we are not subject to regulation which could force sales at undesirable points in time. The broad support for the important elements of the investment strategy enables us to maintain a strategy ... We have rules for rebalancing the equity portion of the fund. This means that the equity portion of the fund and the benchmark is also maintained in periods when the equity prices have fallen a great deal. Maintaining the equity portion is important because the 60% allocation in the benchmark is a tradeoff between expected return and acceptance of risk, so we should maintain that. Rebalancing may also contribute to increasing the return of the fund through the exploitation of variations in equity risk premium. Over time rebalancing has served us well. We just reviewed our rebalancing rule, and calculation shows that it has added .4% to .5% of annual returns since 1998. The rebalancing rules have a number of parameters — it has a lot of things that we can twist and turn on" (Andersson, Ang, Haugerud, Ithurbide & Orr (2012). 
$\\[0.1in]$

To solve such problems, researchers have tried to uncover invariant properties in the distribution of price changes observed over time increments. More than 150 years ago Regnault (1863) observed that the standard deviation of a price change over the time interval $\Delta t$ scales as the square root of $\Delta t$. This observation motivated Bachelier (1900) to formalize the definition of the Brownian motion as a possible model of a stock price. Bachelier posited that price changes are Gaussian, $\mathrm{i.i.d.}$. Some 60 years later, Mandelbrot discerned that price changes have much thicker tails than the Gaussian distribution permits (Mandelbrot (1963)), and proposed to replace the Brownian motion with another family of scale-invariant processes with independent increments - the stable processes of Levy (1924). When $p \left( t \right)$ is the natural log of a stock price or exchange rate, if $p \left( t \right)$ follows a Brownian or Levy process, the distributions of returns $p \left( t + \Delta t \right) - p \left( t \right)$ observed over various horizons $\Delta t$ can be obtained from each other via linear rescaling. Self-similarity thus approximates (somewhat crudely) price changes. Both Brownian motion and Levy processes assume that price changes are independent. Since Engle (1982) there is substantial evidence that the size of price changes, $\mid p \left( t + \Delta t \right) - p \left( t \right) \mid$, is persistent. Subsequently we know that for many time series the size of price changes is a long-memory process characterized by a hyperbolically declining autocorrelogram (see, for example, the ground-setting studies by Ding, Granger & Engle (1993), and Dacorogna, Muller, Nagler, Olsen & Pictet (1993)).
$\\[0.1in]$

Since the mid-1990's, researchers have uncovered different forms of scale-invariance in financial returns, based on multifractal moment-scaling. Early studies by Ghashgaie, Breymann, Peinke, Talkner & Dodge (1996), Mandelbrot, Calvet & Fisher (1997a, 1997b, 1997c), and Calvet & Fisher (2002) found evidence that the moments of the absolute price changes, $\mathbb{E} \left( {\mid p \left( t + \Delta t \right) - p \left( t \right) \mid}^q \right)$, scale as power functions of the horizon $\Delta t$. This observation motivated the construction of the first family of multifractal diffusions (Mandelbrot, Calvet & Fisher (1997a, 1997b, 1997c), Calvet & Fisher (2001), and Bacry, Delour & Muzy (2001). Scale invariance properties allowed researchers to model price variations using a single data-generating mechanism. This resulted in a broad and still expanding class of models that are extremely parsimonious. A small number of well-identified parameters, combined with testable assumptions on scale-invariance, specify price dynamics at all timescales.
$\\[0.1in]$

Mandelbrot advocated that financial prices should be modeled by a broader class of stochastic processes. The real-valued process $\{p \left( t \right) ; t \in \mathbb{R}_+ \}$ is regarded to be self-similar with index $H$ if for all $c > 0$, $n > 0$, and $t_1, t_2, ..., t_n \in \mathbb{R}_+$, the vector $\left( p \left( c t_n \right) \right)$ has the same distribution as $\left( c^H p \left( t_1 \right), c^H p \left( t_2 \right), ..., c^H p \left( t_n \right) \right)$. This may be more concisely written thus:
$\\[0.1in]$

\begin{equation}
    \left(p \left(c t_1 \right),  \left(c t_2 \right), ..., \left(c t_n \right) \stackrel{d}{=} c^H p \left( t_1 \right),  c^H p \left( t_2 \right), ..., c^H p \left( t_n \right) \right)
\end{equation}

$\\[0.1in]$
The Brownian motion is self-similar with $H = \frac {1} {2}$. Levy stable processes are self-similar with independent increments and Paretian tails: 
$\\[0.1in]$

\begin{equation}
    \mathbb{P} \{ \mid p \left( t + \Delta t \right) - p \left( t \right) \mid > x \} \sim K_{\alpha} \Delta t x^{- \alpha}
\end{equation}

$\\[0.1in]$
as $x \rightarrow + \infty$, where $\alpha = \frac {1} {H} \in \left(0; 2 \right)$ and $K_{\alpha}$ is a positive constant.
$\\[0.1in]$

Levy processes have thicker tails than the Brownian motion and are more likely to accommodate large price changes (Mandelbrot (1963, 1967)). But Levy processes have infinite variances, a condition subsequently contradicted by empirical evidence for many time series in the 1970's and 1980's (for example, Blattberg & Gonedes (1974) and Fielitz & Rozelle (1982)). And infinite variances of course present fundamental theoretical difficulties versus the large body of portfolio selection and asset pricing literature based upon the trade-off between mean and variance (see, for example, Campbell, J. Y. & Viceira, L. (2002), and/or the preceding classics by Markowitz (1952), Tobin (1958), Sharpe (1964), and Merton (1992)).
$\\[0.1in]$

Fractional Brownian motions represent another important class of self-similar processes. Mandelbrot (1963, 1965, 1967) presented a stationary process, fractional Gaussian noise which could be used as a time-series model, exhibiting both the Hurst effect and 1/f noise. The fractional Gaussian noise process is the formal differential of H-self similar fractional Brownian motion, and was subsequently developed by Mandelbrot and Van Ness (1968). Fractional Brownian motion's increments are stationary, normally distributed, and strongly dependent. Their autocorrelation declines at the hyperbolic rate:
$\\[0.1in]$

\begin{equation}
    \mathrm{Cov} \left( r_{t + n} ; r_t \right) \sim H \left( 2 H - 1 \right) n^{2H - 2} \; \; \; \; \mathrm{as} \; n \rightarrow \infty
\end{equation}

$\\[0.1in]$
where $r_t = p \left( t \right) - p \left( t - 1 \right)$ denotes the return on a time interval of unit length. The fractional Brownian motion rarely represents a practical model of asset prices however, because the unconditional distributions are Gaussian. Long memory in returns is inconsistent with arbitrage-pricing in continuous time, a finding reported since at least Maheswaran & Sims (1993). The applicability of self-similar processes to finance is also limited by another common shortcomming. By definition, returns observed at various frequencies have identical distributions up to a scalar renormalization $p \left(t + \Delta t \right) - p \left(t \right), \stackrel{d}{=} {\left( \Delta t \right)}^H p \left( 1 \right)$. Most financial time series, however, are not exactly self-similar, but exhibit thicker tails, and for this reason cannot be fully satisfactory models of asset returns.
$\\[0.1in]$

Following Mandelbrot (1963, 1967), several studies documented Paretian tails in financial series, although generally with a tail index larger than 2 and therefore a finite variance. Early evidence (for example, see again Blattberg & Gonedes (1974) and Fielitz & Rozelle (1982)) was parametric. In the 1970's, statisticians developed precise techniques for the nonparametric estimation of the tail index of a distribution (Hill (1975)). This in turn led to several studies showing that financial series have finite tail indexes that are most often strictly larger than 2 (for example, Jansen & DeVries (1991), Koedijk & Kool (1992), and Gabaix (2009)). 
$\\[0.1in]$

In the 1990's researchers started to demonstrate the strong persistence in volatility (for example, Ding, Granger & Engle (1993), and Dacorogna, Muller, Nagler, Olsen & Pictet (1993, 2001)). Long memory was often defined by a hyperbolic decline in the autocovariance function as the lag goes to infinity. Thus, consider that, for every moment $q \geq 0$ and every integer $n$, $\rho_q \left( n \right) = \mathrm{Corr} \left( {\mid r_t \mid}^q , {\mid r_{t + n} \mid}^q \right)$, denotes the autocorrelation in levels. The asset therefore exhibits long memory in the size of returns if $\rho_q \left( n \right)$ is hyperbolic in $n$.
$\\[0.1in]$

The features I have just described can be clearly seen in the data used in the empirical study of this paper. Figure 4 (see section '5. Data' below) shows HF log returns for two stocks: Apple (Apple Inc., NASDAQ, ticker: AAPL) and JPM (JPMorgan Chase & Co., NYSE, ticker: JPM). HF log returns cover the period 01-01-2018 to 31-03-2020. Figure 4 also shows HF log returns over the same period for the EURUSD exchange rate. A benchmark JPYUSD (DEXJPUS) daily log return time series from 01-06-1973 to 28-06-2002 is also shown in the panel. All four time series exhibit both fat tails and volatility clustering at different time scales, including over periods that traverse large parts of the overall series, as occurs in the presence of long memory. Figure 5 shows the same features from a different perspective, by superimposing log returns upon prices for AAPL. This graphic nicely illustrates volatility clustering, which is clearly seen as the amplitude of returns changes frequently from larger to smaller.
$\\[0.1in]$

By the mid-1990's, the combination of fat tails and long memory in volatility led researchers to consider that asset prices may exhibit multifractality. This in turn suggested that financial series should exhibit multifractal moment-scaling, a generalization of self-similarity defined by:
$\\[0.1in]$

\begin{equation}
    \mathbb{E} \left( {\mid p \left( t + \Delta t \right) - p \left( t \right) \mid}^q \right) = c_q {\left( \Delta t \right)}^{\tau \left( q \right) + 1}
\end{equation}

$\\[0.1in]$
which holds for every (finite) moment $q$ and time interval $\Delta t$. A self-similar process satisfies this relation, with $\tau \left(q \right) = H q - 1$. The process $p$ is said to be strictly multifractal if equation 3.4 holds for a strictly concave function $\tau \left( q \right)$. Strict multifractal moment scaling has been shown in a number of studies, for example Calvet & Fisher (2002), Mandelbrot, Calvet & Fisher (1997c), and Ghashgaie, Breymann, Peinke, Talkner & Dodge (1996). Saichev & Sornette (2006) studied the multifractal properties of a class of continuous stochastic processes consistent with equation 3.4, finding that the differences between exact multifractal random walks and exact multifractal scaling could not be distinguished from approximate scaling laws given the noise and the limited ranges used in their experimental and numerical set ups. Filimonov & Sornette (2010) subsequently exposited a self-excited multifractal process and demonstrated that it is consistent with equation 3.4 and exhibits multifractal scaling over a large intermediate range of scales. 
$\\[0.1in]$

Another class of stochastic processes consistent with equation 3.4 was provided by the multifractal model of asset returns (MMAR), the first example of a multifractal diffusion (Mandelbrot, Calvet & Fisher (1997a, 1997b, 1997c), Calvet & Fisher (2002)). MMAR built upon multifractal measures (see Mandelbrot (1974)), constructed via an iterative random reallocation of mass within a time interval. Calvet and Fisher provided the example of the binomial measure on $\left[ 0, 1 \right]$, which they derived as the limit of a multiplicative cascade. 

MMAR's formulation is important to the binomial MSM specification I use later in the empirical study. Thus, consider the uniform probability measure $\mu_0$ on the unit interval. Let $M$ denote a binomial distribution taking the high value $m_0 \in \left[\frac {1} {2}, 1 \right]$ or the low value $1 - m_0$ with equal probability. In the first step of the cascade, two independent values $M_0$ and $M_1$ may be drawn from the binomial. Define a measure $\mu_1$ by uniformly spreading the mass $M_0$ on the left subinterval $\left[ 0, \frac {1} {2} \right]$, and the mass $M_1$ on the right subinterval $\left[\frac {1} {2}, 1 \right]$. The density of $\mu_1$ is a step function. In the second stage of the cascade, draw four independent binomials $M_{0, 0}$, $M_{0, 1}$, $M_{1, 0}$, $M_{1, 1}$. Split the interval $\left[ 0, \frac {1} {2} \right]$ into two subintervals of equal length: the left subinterval $\left[ 0, \frac {1} {4} \right]$ is allocated a fraction $M_{0, 0}$ of $\mu_1 \left[ 0, \frac {1} {2} \right]$; whilst the right subinterval $\left[ \frac {1} {4}, \frac {1} {2} \right]$ receives the fraction $M_{0, 1}$. Apply a similar procedure to $\left[\frac {1} {2}, 1 \right]$, to obtain a measure $\mu_2$, such that:
$\\[0.1in]$

\begin{equation}
    \mu_2 \left[ 0, \frac {1} {4} \right] = M_0 M_{0, 0}, \; \; \mu_2 \left[ \frac {1} {4}, \frac {1} {2} \right] = M_0 M_{0, 1}, \; \; \mu_2 \left[ \frac {1} {2}, \frac {3} {4} \right] = M_1 M_{1, 0}, \; \; \mu_2 \left[ \frac {3} {4}, 1 \right] = M_1 M_{1, 1}
\end{equation}

$\\[0.1in]$
Iteration generates an infinite sequence of random measures $\left( \mu_k \right)$ that weakly converges to the binomial measure $\mu$.
$\\[0.1in]$

Now consider the interval $\left[ t, t+2^{-k} \right]$, where $t = \sum^k_{i=1} \eta_i 2^{-i}$ and $\eta_1, \eta_2, ..., \eta_k \in \{0, 1\}$, and let $\varphi_0$ and $\varphi_1$ denote the relative frequencies of 0's and 1's in $\left( \eta_1, \eta_2, ..., \eta_k \right)$. The measure of the dyadic interval is then:
$\\[0.1in]$

\begin{equation}
    \mu \left[ t, t + 2^{-k} \right] = M_{\eta_1}, M_{\eta_1, \eta_2}, ..., M_{\eta_1, ..., \eta_k} \Omega
\end{equation}

$\\[0.1in]$
where $\Omega$ is a random variable determined by the change in the mass generated by stages $k + 1, k + 2, ..., \infty$ of the cascade. Furthermore:
$\\[0.1in]$

\begin{equation}
    \mathbb{E} \left( \mu {\left[ t, t + 2^{-k} \right]}^q \right) = {\left[ \mathbb{E} \left( M^q \right) \right]}^k = {\left( \Delta t \right)}^{\tau_{\mu} \left( q \right) + 1}
\end{equation}

$\\[0.1in]$
where $\Delta t = 2^{-k}$ and $\tau \left( q \right) = - \log_2 \left( E \left( M^q \right) \right) - 1$. The moments of the limiting measure of a dyadic interval of length $\Delta t$ are therefore a power of $\Delta t$.
$\\[0.1in]$

The MMAR extended multifractals from measures to diffusions. An asset price could be specified by compounding a Brownian motion with an independent random time-deformation, thus:
$\\[0.1in]$

\begin{equation}
    p \left( t \right) = p \left( 0 \right) + B \left[ \theta \left( t \right) \right]
\end{equation}

$\\[0.1in]$
where $\theta$ is the cumulative distribution of the multifractal measure $\mu$. The price process inherits the moment-scaling properties of the measure, so that $\mathbb{E} \left( {\mid p \left( t + \Delta t \right) - p \left( t \right) \mid}^q \right) = {\left( \Delta t \right)}^{\tau_{\mu} \left( q / 2 \right) + 1}$ on any dyadic interval $\left[ t, t + \Delta t \right]$. These moment restrictions represent the basis of estimation and testing (see Mandelbrot, Calvet & Fisher (1997a, 1997b, 1997c), Calvet & Fisher (2002), plus the subsequent paper by Lux (2008)). MMAR is thus a well-defined stochastic framework for the analysis of moment-scaling. Andersen, Bollerslev, Diebold, & Labys (2001) and Barndorff-Nielsen & Shephard (2004) produced related research on power variation, which interpreted return moments at various frequencies in the context of traditional jump-diffusions. However, whilst MMAR has many appealing properties, it has been limited in implementation by two features of the underlying measure. Firstly, the recursive reallocation of mass on an entire time-interval does not fit well with standard time series tools. Secondly, the limiting measure contained a residual grid of instants makes it non-stationary.
$\\[0.1in]$

Calvet & Fisher (2004, 2008b) resolved these difficulties by constructing a fully stationary volatility process that evolves stochastically through time: the Markov-switching multifractal (MSM). The MSM model is a milestone in the multifractal literature, bridging multifractality and regime-switching; I address the MSM model, its empirical estimation and forecasting in detail in section '6.2 Markov-switching multifractal'.
$\\[0.1in]$

The MSM construction has also been applied in continuous time (see Calvet & Fisher (2001, 2008a, 2008b) and Saichev & Sornette (2006)) and the continuous-time MSM set up can be summarized as follows. Consider a filtration $\{\mathcal{F_t}\}$ on the probability space $\left( \Omega, \mathcal{F}, \mathbb{P} \right)$. The Markov state vector
$\\[0.1in]$

\begin{equation}
    M_t = \left( M_{1, t}; M_{2, t}; ..., M_{{\bar k}, t} \right) \in \mathbb{R}^{\bar k}_+
\end{equation}

$\\[0.1in]$
is now defined for all $t \in \mathbb{R}_+$. With the Markov state $M_t$ at data $t$, the dynamics over an infinitesimal interval may be defined as follows. For each $k \in \{ 1, 2, ..., {\bar k} \}$, a change in $M_{k, t}$ may be triggered by a Poisson arrival with intensity $\lambda_k$. The component $M_{k, t + dt}$ is drawn from a fixed distribution $M$ if there is an arrival, and otherwise remains at its current value: $M_{k, t + dt} = M_{k, t}$. The construction may be summarized thus:
$\\[0.1in]$

\begin{equation}
    M_{k,t + dt} =
    \begin{cases}
    \text{ drawn from distribution M }  & \text{ with probability } \lambda_k dt \\
    M_{k,t} & \text { with probability }  1 - \lambda_k dt
    \end{cases}
\end{equation}

$\\[0.1in]$
The Poisson arrivals and new draws from $M$ are independent across $k$ and $t$. The sample paths of a component $M_{k, t}$ are càdlàg, that is, right-continuous with a limit point to the left of any instant. The arrival intensities may be specified thus:
$\\[0.1in]$

\begin{equation}
    \lambda_k = \lambda_1 b^{k - 1}, \; \; \; \; k \in \{1, 2, ..., \bar k \}
\end{equation}

$\\[0.1in]$
The parameter $\lambda_1$ determines the persistence of the lowest frequency component, whilst $b$ determines the spacing between component frequencies. The price process follows an Ito diffusion thus:
$\\[0.1in]$

\begin{equation}
    \frac {dP_t} {P_t} = \mu dt + \sigma \left( M_t \right) dZ_t
\end{equation}

$\\[0.1in]$
where $\sigma \left(M_t \right)$ follows 
$\\[0.1in]$

\begin{equation}
    \sigma \left( M_t \right) = \bar \sigma {\left( \prod_{k=1}^{\bar k} M_{k, t} \right)}^{\frac {1} {2}}
\end{equation}

The MSM framework offers considerable flexibility and has thus been expanded in several directions. Calvet, Fisher & Thompson (2006) introduced a multivariate model of asset returns, capturing the correlation both in levels and in volatility of the returns on several financial assets. They demonstrated that multivariate MSM preserved the tractability of univariate MSM, capturing well the joint dynamics of asset returns and providing accurate forecasts of the value-at-risk (VaR) of a portfolio of assets. Idier (2008, 2009) developed an extension of bivariate MSM (the MSM Dynamic Conditional Correlation model (MSMDCC)) that incorporates dynamic correlation in the Gaussian innovations. MSMDCC combined the multi-frequency structure of bivariate MSM with the flexible correlation of Engle's (2002) dynamic conditional correlation model. The likelihood and Bayesian filter of MSMDCC are available analytically and Idier (2009) showed that it outperforms standard MSM and DCC set ups in- and out-of-sample. Alausa (2014) extended the bivariate MSM by incorporating a vector error correction mechanism, to yield the Vector Error Correction MSM (VEC-MSM). VEC-MSM used an error correction mechanism in the mean equation to incorporate long-run relationships between spot and futures prices, which is combined with the MSM volatility process. This research demonstrated significant risk reduction (via VaR metrics) in- and out-of-sample on commodities, foreign exchange and equities data. Lux & Morales-Arias (2009) extended the MSM set up with Student-t innovations (MSM-t), demonstrating improved performance versus (G)-ARCH-class benchmarks. Wang, Wu & Yang (2016) studied crude oil return volatility with a binomial MSM, obtaining significant performance improvements, and Segnon, Lux & Gupta (2015) forecast carbon dioxide (CO2) price volatility using a Lognormal MSM. 
$\\[0.1in]$

Zikes, Barunik & Shenai (2018) and Chen, Diebold & Schorfheide (2013) "independently and in parallel" introduced MSM models of the time interval between two trades on a given security. The resulting MSM duration (MSMD) models were shown to capture the key features of financial market inter-trade durations: long memory dynamics and highly dispersed distributions. Zikes, Barunik & Shenai compared the performance of MSMD to competing short- and long-memory duration models, with MSMD dominating in- and out-of-sample. Aldrich, Heckenbach & Laughlin (2014) subsequently extended upon the characterization of inter-trade durations in order to provide a better characterization of HF durations.
$\\[0.1in]$

Several researchers have studied the pricing implications of multifractal risk. Calvet & Fisher (2007) developed a discrete-time model of stock returns in which the volatility of dividend news followed a MSM. The resulting variance of stock returns is shown to be substantially higher than the variance of dividends, in line with the data. Calvet & Fisher also show that the MSM dividend specification improved upon the classic Campbell & Hentschel (1992) model, which generated more modest amplification effects with a GARCH dividend process. Calvet & Fisher (2007) also investigated the dynamics of returns when the agent is not fully informed about the state $M_t$, but must sequentially learn about it from dividends and other signals. The implied return process in this study exhibited substantial negative skewness, again in line with the data. Hainaut (2012) replaced the constant volatility of the Brownian motion in the Hull-White model with a MSM process. Motivated to capture excess kurtosis and skewness in rates distributions, this study showed that fractal volatility provides a more realistic model of the behavior of short term rates (compared to Hull-White (1990)).
$\\[0.1in]$

Multifractal modeling has been applied to option pricing. Calvet, Fearnley, Fisher, & Leippold (2015) introduced an extension of MSM that accounts for the variation in skewness and term-structure of option data. In this study, jumps to the return process were triggered by changes in lower-frequency volatility components, so that a 'leverage effect' is generated by a negative correlation of HF innovations to returns and volatility. Using index and option maturities and strikes data, latent volatility components were shown to enable the model to dynamically fit a wide range of option surfaces.
$\\[0.1in]$

Beyond extension of the MSM framework, a broader multifractal literature has been catalyzed by Calvet & Fisher's work and the MSM.  Alternate multifractal formulations have been put forward, such as the Multifractal Random Walk (MRW) of Bacry, Delour & Muzy (2001), extended by Bacry, Kozhemyak & Muzy (2008), and the Self-Excited Multifractal (SEMF) (Filimonov & Sornette (2010)). There are several expository works that test for multifractality in data and confirm multiscaling (for example, Liu, Matteo & Lux (2007), and Saichev & Sornette (2006)). Alternative estimation and optimization methods have been studied. Lux (2008) proposed an alternative GMM estimator, motivated by less computationally demanding linear forecasting. In a subsequent paper (Lux (2018)) studied the performance of two numerical optimization methods, the Iterated Filtering I and II algorithms of Ionides, et al. (2006, 2015). In this research Lux found GMM less efficient than maximum likelihood.  GMM with linear forecasts was typically observed to be disadvantaged against ML-based optimal forecasts. Lux, Morales-Arias & Sattarhoff (2011) forecast realized volatility, again using best linear forecasts derived from GMM estimation.
$\\[0.1in]$

# Microstructure noise

In classic models of asset pricing $X$, the log of an asset price, is assumed to follow an Ito semimartingale. This model originates with Bachelier (1900), Osborne (1959), Cootner (1964), Samuelson (1965), and Mandelbrot (1966); the current paradigm was introduced and formalized using stochastic analysis by Merton in a series of seminal contributions (collectively, see Merton (1992). For an asset pricing model to avoid arbitrage opportunities, asset prices must follow semimartingales (see, for example, Ait-Sahalia & Jacod (2014), Robert & Rosenbaum (2012), Delbaen & Schachermayer (1994), and Harrison & Pliska (1981)). A semimartingale can be decomposed into the sum of a drift, a continuous local martingale, and a discontinuous part. The continuous component can be scaled by a stochastic volatility process, which may be correlated with the asset price, and may itself be a semimartingale. From a continuous-time financial econometrics perspective, methods were initially developed to estimate models using low frequency data only (again see Ait-Sahalia & Jacod (2014); also  Ait-Sahalia (1996) and Hansen & Scheinkman (1995)). Subsequent availability of high frequency data then enabled the development of methods with the objective of estimating the volatility component using estimators such as RV (Liu, Patton & Sheppard (2015) and Andersen, Bollerslev & Diebold (2010) provide detailed reviews). 
$\\[0.1in]$

However, at high frequency, the properties of volatility estimators may be radically altered by the presence of "microstructure noise". This insight stemmed from the development of methods to make high frequency volatility estimators more robust. Initially these methods focused exclusively on models with continuous sample paths. They include, in chronological order, the two scales realized volatility (TSRV) estimator of Zhang, Mykland & Ait-Sahalia (2005), the maximum likelihood approach of Ait-Sahalia, Mykland & Zhang (2005), the multi-scaled realized volatility (MSRV) of Zhang (2006), the realized kernels of Barndorff-Nielsen, Hansen, Lunde & Shephard (2009, 2008, 2007), the preaveraging approach of Jacod, Li, Mykland, Podolski & Vetterd (2009), and the quasi-maximum likelihood estimator of Xiu (2010). This literature is augmented by more recent work from Figueroa & Wu (2020), Andersen, Thyrsgaard, & Todorov (2018), Bollerslev, Li, & Xue (2018), Bollerslev, Hood, Huss & Pedersen (2017), along with earlier work by Ait-Sahalia, Mykland & Zhang (2012, 2011), Gatheral & Oomen (2010), Li, Mykland, Renault, Zhang, & Zheng, (2010), Kalnina & Linton (2008), Cheng & Sen (2008), Li & Mykland (2007), Fan & Wang (2007), Hansen & Lunde (2006), Bandi & Russel (2006), and Delattre & Jacod (1997).
$\\[0.1in]$

RV provides a consistent estimate of the quadratic variation of asset prices when prices follow a continuous stochastic model and are directly observed (see, for example, the pioneering papers of Barndorff-Nielsen & Shephard (2002) and Andersen, Bollerslev, Diebold & Labys (2003)). This is compelling and it makes RV a very simple and natural volatility estimator. RV is also, as noted by Ait-Sahalia & Yu (2009), possessed of the most desirable properties. For instance, in the context of parametric volatility models, RV is not only the least squares estimator or the method of moments estimator with the sample variance as the moment function, but also the maximum likelihood estimator. The asymptotic properties of RV are striking when sampling occurs at an increasing frequency that is a realistic approximation to what is observed with, for example, the nano-second NYSE Trades and Quotes and HF FX data that are source data for the empirical study in this paper. In particular, fully observing the sample path of an asset can be shown, in the limit, to perfectly reveal the volatility of that path. This result is nonparametric in nature, in that the estimator will converge to the quadratic variation of the process, and holds in great generality for semimartingales. However, in an HF empirical setting, things are not necessarily as simple as these asymptotic results suggest. Controlling for microstructure noise that is prevalent at high frequencies is a key issue.
$\\[0.1in]$

Microstructure noise captures a wide range of frictions inherent in the information flows associated with the trading process; an insight that goes back to Roll (1984) and Hasbrouck (1993). Ait-Sahalia & Jacod (2014) and Menkveld (2016), for example, describe the frictions as bid-ask bounce, discreteness of price changes, differences in trade sizes or informational content of price changes, gradual response of prices to a block trade, strategic component of the order flow, and inventory control effects. Bacry, Delattre, Hoffmann & Muzy (2013) describe microstructure noise as strong mean reversion effects at small scales. Theoretically, since at least Merton (1980), it is known that spot volatility may be inferred perfectly if the asset price follows a diffusion process and a continuous record of prices is available. But, as shown consistently in the literature, by practitioners, and indeed in the empirical study of this paper, practical implementation presents significant challenges. 
$\\[0.1in]$

Firstly, even within the nano-second time-frames of current market infrastructures and operations, prices are yet only observed at intermittent and discrete points in time. This induces discretization errors. Secondly, and perhaps more importantly, recorded quotes and prices do not reflect direct observations of a frictionless diffusive process. The order book is the dominant system now in financial markets. It matches orders on a price-time priority and keeps track of all previously submitted buy and sell limit orders on a discrete grid of prices. Prices therefore are quoted on this discrete price grid with a Bid-Ask spread, so that markedly different prices may be quoted simultaneously by competing market makers due to a broad range of factors (an expanding literature has addressed this observation: see for example Stoikov (2020), Dacorogna (2018), Bollerslev, Li & Xue (2018), Lehalle & Mounjid (2018), Lehalle & Laruelle (2014), Robert & Rosenbaum (2012), Andersen, Bollerslev & Meddahi (2011), and  Gatheral & Oomen (2010)). Complications include heterogeneous beliefs (see for example, Rousseau, Boco & Germain (2016) and Diebold & Strasser (2013)), information (Andersen, Bollerslev & Meddahi (2011)), transactions costs (Amihud, Mendelson & Pedersen (2005)), inventory positions (Andersen, Bollerslev & Meddahi (2011)), inter-trade duration (Zikes, Barunik & Shenai (2018), Aldrich, Heckenbach & Laughlin (2014), Chen, Diebold & Schorfheide (2013), Robert & Rosenbaum (2012), Sun & Engle (2007), and Engle (2000)),  volume (Bollerslev, Li & Xue (2018), Ait-Sahalia, Mykland & Zhang (2011), Griffin & Oomen (2011)), and liquidity (Ait-Sahalia & Yu (2009), Ait-Sahalia & Mykland (2009)). Variously at times, often collectively, these complications are labeled microstructure noise. Consequently, any observed price may be regarded to not represent a unique market price but instead an underlying ideal price that is confounded by an error term reflecting the impact of this 'noise'. The biggest contributory factor to microstructure noise has been consistently shown to be Bid-Ask bounce (Lehalle & Laruelle (2014), Ait-Sahalia, Mykland & Zhang (2010), Almgren (2009), Barndorff-Nielsen, Hansen, Lunde & Shephard (2009), Dacorogna, Gencay, Muller, Olsen, & Pictet (2001)). 
$\\[0.1in]$

The importance attributed to Bid-Ask bounce by the literature with regard to microstructure noise may be explained as follows. At very high frequencies (i.e. tick data) underlying volatility may be zero. But, buyers and sellers are still present and consequently prices are observed at Bid or Ask, rather than at mid-price. Because of that, if the econometrician chooses a classical quadratic variation estimate for squared volatility, even with an underlying volatility of zero, she will measure a lot of time ${\left(P_{Ask} - P_{Bid} \right)}^2$.  She may model this with additive noise, $\epsilon_t$, so that, when the mid-price between $0$ and $T$ follows a discretized arithmetic Brownian motion, $P_{\left(k+1 \right)\delta} = P_{k\delta} + \sigma \sqrt{\delta} \xi_{k+1}$, the price she observes at $k\delta$ is $P_{k\delta} + \epsilon_k$. Empirically, $\epsilon_k$ has been shown to approximate to half a bid-ask spread (see for example Lehalle (2012) and Almgren (2009)).
$\\[0.1in]$

Consequently, sampling $\left[ 0, T \right]$ in $N$ intervals $\left( \delta = \frac {T} {N} \right)$, the expectation of the quadratic variation, which should be an estimate of $\int_0^T \sigma_t^2 dt$, may be described:
$\\[0.1in]$

\begin{align}
    U \left(N \right) &= \mathbb{E} \left( \sum_{n=1}^N {\left( P_{n\delta} + \epsilon_n - \left( P_{\left(n-1 \right)\delta} + \epsilon_{n-1} \right) \right)}^2 \right) \\
                      &= \sum_{n=1}^N \mathbb{E} {\left( P_{n \delta} - P_{\left(n-1 \right) \delta} \right)}^2 + \mathbb{E} {\left( \epsilon_n - \epsilon_{n-1} \right)}^2
\end{align}

$\\[0.1in]$
where $U$ is noise, following Barndorff-Nielsen, Hansen, Lunde & Shephard (2007, 2008, 2009).
$\\[0.1in]$

Consequently, the estimate may now be described:
$\\[0.1in]$

\begin{align}
    U \left(N \right) \rightarrow \int_0^T \sigma_t^2 dt + 2Nv
\end{align}

$\\[0.1in]$
where $N \rightarrow +\infty$. Thus, in this setting, the estimate of squared volatility increases linearly with the sampling rate, driven only by the microstructure noise (with variance $v$).
$\\[0.1in]$

Early approaches to the problem of microstructure noise in the empirical literature consisted of discarding the data sampled at the highest frequencies, as the 'noisiest', in order to sample instead at 15 minute, 30 minute, or even longer intervals (see Andersen, Bollerslev, Christoffersen & Diebold (2006), Barndorff-Nielsen & Shephard (2007), Meddahi, Mykland & Shephard (2011) for comprehensive reviews of these early approaches). To quote Zhang, Mykland, & Ait-Sahalia (2005), “it is difficult to accept that throwing away data, especially in such quantities, can be an optimal solution”. More recent approaches therefore explicitly incorporate microstructure noise into the analysis, so that estimators make use of all the data, or as much data as the econometrician may choose to use. These methods make it possible to decompose the total observed variance into a component attributable to the fundamental price signal and one attributable to the microstructure noise. These estimators may also, in some settings, produce consistent estimates of the magnitude of the market microstructure noise at high frequencies, thereby producing a decomposition of total asset return volatility into fundamental and noise components. At least two distinct approaches to modeling this noise component may be encountered. 
$\\[0.1in]$

Firstly, and most common, is the use of an additive noise that is independent of the efficient price (studies following this approach include  Ait-Sahalia, Mykland & Zhang (2010), Zhang (2006), Ait-Sahalia, Mykland & Zhang (2005), Bandi & Russel (2005), and Zhang, Mykland & Ait-Sahalia (2005)). An additive endogenous component is also considered in Barndorff-Nielsen, Hansen, Lunde & Shephard (2009), through a linear combination of past returns, and in Kalnina & Linton (2008) through the Brownian motion driving the price. A potential limitation of these approaches is that they do not allow for prices discreteness. A second, alternative approach is where microstructure noise is proposed as cases of a contamination of the efficient price through a Markov kernel (this approach is followed in studies by Jacod, Li, Mykland, Podolski, & Vetterd (2009), and Li & Mykland (2007)), and as a rounding error (Delattre & Jacod (1997) and Li & Mykland (2008)). The microstructure noise is respectively partially and fully endogenous in these approaches. 
$\\[0.1in]$

Of the two approaches, additive microstructure noise is the dominant paradigm. It may be explained as follows. Assume a time horizon $\left[ 0, 1 \right]$ and consider a price $P_t$ observed at time $t = \frac {i} {n}, i = 0, ..., n$, taking the form:
$\\[0.1in]$

\begin{align}
    \log{\left(P_{i / n} \right)} = \log{\left(X_{i / n} \right)} + \epsilon^n_i
\end{align}

$\\[0.1in]$
where $X$ is a continuous-time semimartingale, representing an efficient price, and $\epsilon^n_i$ is a microstructure noise term. 
$\\[0.1in]$

The prefix "additive" implies that in this setting the econometrician focuses upon the properties of the noise process $\epsilon_t = P_t - X_t$. In this way, she obtains quite straightforwardly, firstly data that is close to a semimartingale in lower frequencies (i.e. $P_{i / n} - P_{j / n} \approx X_{i / n} - X_{j / n}$ as soon as $\left(i - j\right) / n$ is large enough); but secondly, data very different from that of a semimartingale in higher frequencies, the noise term becoming increasingly important as the data frequency increases.
$\\[0.1in]$

The model [X - the $\log{\left(P_{i / n} \right)}$ formula immediately above] setting enables important desirable features for the observed price and microstructure noise. Firstly and perhaps most obviously, the separation of the price and the microstructure noise is straightforward. However, the econometrician observes discrete prices that are incompatible with observations of a semimartingale at exogenous samping times. The important impact of this discretization effect on various statistical procedures has been emphasized since the mid-1980's (for example Ball (1988), Cho & Frees (1988), and Gottlieb & Kalay (1985)). Secondly, the econometrician observes quick oscillations of the transaction price between the Bid-price and the Ask-price, this being of course  the Bid-Ask bounce.
$\\[0.1in]$

# Data

## Preparation

Careful data cleaning is one of the most important aspects of volatility estimation with high frequency data (see Barndorff-Nielsen, Hansen, Lunde & Shephard (2009)) and it is not always straightforward to construct a time series of interest from raw high frequency data. The methods I employ to prepare the data and compute log returns for this empirical study therefore are not trivial. I follow the data cleansing procedures set out in Bollerslev, Hood, Huss & Pederson (2017), Barndorff-Nielsen, Hansen, Lunde & Shephard (2009) and Brownlees & Gallo (2006), which are in turn consistent in many respects with Dacorogna, Gencay, Muller, Olsen, & Pictet (2001). Variants of these procedures are also followed in several other important papers, including Bollerslev, Li & Xue (2018), Ait-Sahalia & Jacod (2012), Ait-Sahalia, Mykland & Zhang (2010), Barndorff-Nielsen, Hansen, Lunde & Shephard (2007, 2008), and Hanson & Lunde (2006). 
$\\[0.1in]$

The consequences of data handling procedures upon subsequent statistical analysis may be significant (Brownlees & Gallo (2006)). For example, Hansen & Lunde (2006) show that removing large numbers of observations can improve volatility estimation. This result may appear counterintuitive, but the reasoning is quite straightforward: An estimator that makes optimal use of all data will typically put high weight on accurate data and be less influenced by the least accurate observations. Barndorff-Nielsen, Hansen, Lunde & Shephard (2009) develop a good analogy for this using the generalized least squares estimator in the classical regression model. Conversely, the precision of the standard least squares estimator can deteriorate when relatively noisy observations are included in the estimation. Thus the inclusion of poor quality observations can cause more harm than good to the least squares estimator. Realized volatility kernels and related estimators "treat all observations equally" (Barndorff-Nielsen, Hansen, Lunde & Shephard (2008)) so that a few outliers can severely influence these estimators. By contrast, in my research I am concerned with the extent to which the complex, non-linear, stochastic MSM with its state space architecture is able to treat observations differently and perform better as a result.  There is therefore an important balance to be struck in the process of data preparation between too little data cleansing and too much. As a result, a great deal of raw data analysis and trial-and-error experimentation has accompanied the data preparation process described subsequently in this section.
$\\[0.1in]$

My equity data is high frequency tick-by-tick data for two stocks: Apple (Apple Inc., NASDAQ, ticker: AAPL) and JPM (JPMorgan Chase & Co., NYSE, ticker: JPM). The selection of these two particular stocks is ultimately an arbitrary model setting choice. However, Ait-Sahalia & Jacod (2012) show that market microstructure noise is greater in NASDAQ-listed stocks than Dow Jones Industrial Average (DJIA) stocks listed on NYSE. A cursory look at the performance of the stocks over the period of analysis for this research (see Fig 3 for example) suggests prima facie that the behaviour of these two stocks might differ somewhat in ways that are interesting for this research agenda. The data is obtained from the New York Stock Exchange (NYSE) Trade and Quote (TAQ) database. TAQ is an historical archive of daily files containing all trades and quotes for all issues listed and traded on US regulated exchanges for each single trading day. Daily TAQ files thus comprise the entire history of intra-day transactions data for all securities traded on NYSE, Nasdaq and US regional Exchanges (collectively henceforth the Exchange or Exchanges). TAQ consolidates the National Best Bid Offer (NBBO) with quote, trade and other messages from the Exchanges. The NBBO is calculated and disseminated by Security Information Processors (the SIP) as part of the National Market System Plan (NMSP), which is used to process security prices. There are two SIPs responsible for this task: the Consolidated Quotation System gives the NBBO for securities listed on NYSE; and the Unlisted Trading Privileges Quote Data Feed gives the NBBO for securities listed on NASDAQ. The NBBO updates throughout the day with the highest and lowest offers for a security among the Exchanges and market makers. The lowest Ask price and the highest Bid price are displayed in the NBBO and are not required to come from the same Exchange. I obtain TAQ data in the form of compressed historical daily files and extract it with scripts I have written in Python. I have written additional Python scripts to select JPM and AAPL tickers, process the data, analyze, cleanse, consolidate, and prepare it for the computation of WeightedMidPrice log returns. I ultimately consolidate the data into a single time-series and enable my TAQ data processing workflow as a whole with GPU acceleration using CuPy and RAPIDS cuDF in order to effectively manage the computational overhead arising from the very large amounts of raw data involved. The starting point for data preparation is thus 565 daily files, each comprising several gigabytes of data, or millions of underlying transaction (price-quote) records.  The scripts are posted along with other source code for this paper on my github. 
$\\[0.1in]$

It has been reported since French & Roll (1986) that volatility tends to be higher during Exchange trading hours than non-trading hours. Theory underlying the consistency of volatility measures portrays prices as evolving continuously through time. In actuality, of course, most markets close (i.e. weekends and certain holidays), change their trading hours, and sometimes experience "ghost" hours, where liquidity is very poor despite the markets technically being open (Bollerslev, Li & Xue (2018), Foley & Reynolds (2018), Bollerslev, Hood, Huss & Pederson (2017)). Accordingly, I only retain the trading hours for which liquidity is sufficiently high to ensure a reasonable quality of the high frequency data. For the same reason, I filter out non-public exchange trading. My GPU-accelerated TAQ data cleansing and processing workflow comprises the following steps.
$\\[0.1in]$

I focus upon a continuous window of time from when the first Exchange opens until the last Exchange closes in order to cover the whole trading day. Thus, my window covers pre-market (04:00:00 to 09:29:59), market (09:30:00 to 16:00:00), and extended hours (16:00:01 to 20:00:00). Outside of regular market hours (i.e. market (09:30:00 to 16:00:00)), I observe more spurious Bids and Asks. However, in contrast to, for example, Barndorff-Nielsen, Hansen, Lunde & Shephard (2009), I retain pre-market and extended hours activity because, whilst present in the pre-market and extended hours, spurious activity comprises less than 1% of the data. An example of spurious activity outside of the 04:00 to 20:00 window is bad prices. To illustrate, I verify using historical data enquiry functionality on a Bloomberg terminal that 21-Aug-2019 19:27:05.010227335 APPL has Bid 212.05 versus a NBBO entry in the TAQ data of 112.05 that is detected by my scripts. Given the relatively low level of spurious activity in the pre-market and extended hours windows I elect to retain this data at this point in the workflow and deal with it later (i.e. remove the spurious items) via subsequent data cleansing steps. Also, pre-market and extended hours activity is now (i.e. in my period of observation, 2018-2020) significantly deeper than it was at the time of earlier authors' work, so the information in the pre-market and extended hours windows is potentially significant. Entries with a time stamp outside the 04:00 to 20:00 window are deleted, in a manner consistent with literature. I observe through preliminary data analysis that outside of the 04:00 to 20:00 window there are a great many more spurious Bids and Asks (>20% under some test conditions), which I attribute to the fact that no real market is being made.
$\\[0.1in]$

I delete entries where Bid and/or Ask equals zero. When multiple quotes have the same time-stamp, I replace them all with a single entry with the median Bid and median Ask price. I delete entries for which the spread is negative and I delete entries for which the spread is more than 50 times the median spread on that day. In order to remove outliers that are missed by these earlier steps, I delete entries for which the mid-quote deviates by more than 10 mean absolute standard deviations from a rolling centered median (excluding the observation under consideration) of 100 observations (50 observations before and 50 after). By basing the window on observation counts I allow it to expand and contract in clock time depending upon the intensity of trading. The choice of 100 observations is ad hoc (Barndorff-Nielsen, Hansen, Lunde & Shephard (2009), for example, use 50), but based upon a considerable amount of experimentation with different settings.
$\\[0.1in]$

Brownlees & Gallo (2006) use a different approach for removing outliers. These authors remove observation $i$ if the condition  $\mid p_i - \bar p_i \left(k\right)\mid \; >\;  3 s_i \left(k\right) + \gamma$ is true, where $\bar p_i \left(k\right)$ and $s_i \left(k\right)$ denote respectively the $\gamma$-trimmed sample mean and sample standard deviation of a neighbourhood of $k$ observations around $i$, where $\gamma$ is a granularity parameter. I use the median in place of the trimmed sample mean, $\bar p_i \left(k\right)$, and the mean absolute deviation from the median in place of $s_i \left(k\right)$. By not using the sample standard deviation I become less sensitive to runs of outliers, in a manner that generally follows Barndorff-Nielsen, Hansen, Lunde & Shephard (2009).
$\\[0.1in]$

I filter out Bids and Asks that are not originating from a public Exchange. Non-public trading, for example trades done internally by broker dealers (i.e. when a broker dealer crosses two orders internally to create a trade, block trades, and/or when two institutional orders are crossed to create a trade), dark pools (i.e. private Exchanges deemed "dark" because the orders cannot be seen), and Over The Counter (OTC) trades, are included in the TAQ daily files along with associated quote activity. Regulation National Market System (Reg NMS), a set of rules passed by the Securities and Exchange Commission (SEC) in 2005, requires all trades to be reported; thus private trading, trades executed outside of public Exchanges, along with associated quote activity are reported to a Trade Reporting Facility (TRF). TRFs are provided by the Exchanges, which then publish to the SIP. Reg NMS allows up to 10 seconds after execution for a report to be sent to a TRF. This can result in the inclusion on the TAQ at a given instance in time of TRF information that is not within the current NBBO and, as a result, prices and sizes may be unrealistic in proximity to TRF information.
$\\[0.1in]$

Literature is mixed regarding the impact of private trading. O’Hara & Ye (2011) examine effective spreads, realized spreads, execution speed, short-term volatility, and variance ratios and show that fragmentation arising from Reg NMS is associated with lower spreads, faster execution, and prices that are closer to a random walk; however they also find evidence of increased short-term volatility. Conversely, Chung and Chuwonganant (2012) report that Reg NMS has led to increases in quoted and effective spreads, slower execution, and reduced depth, concluding that fragmentation arising from Reg NMS hurts market quality. Foley & Reynolds (2018) and Haslag & Ringgenberg (2016) report similar findings with regard to negative impacts of Reg NMS. Barndorff-Nielsen, Hansen, Lunde & Shephard (2009) report that it is better to analyze Exchanges separately than treat all data as if it was from a single source, and filter TAQ to NYSE entries only in their work.  Hansen & Lunde (2006) also showed that pooling mid-quotes from multiple exchanges can induce additional noise that overshadows the endogenous noise found in single exchange mid-quotes. However, almost a decade later, Bartlett & McCrary (2019), report that delay in information flows onto the SIP from the different Exchanges "is miniscule and becomes smaller every year". These authors show, for example, that SIP-reported NBBO matches a calculated NBBO  without reporting latencies for 97% of observed data. Moreover, they report that this result was generally persistent across the Exchanges.
$\\[0.1in]$

Having prepared the data, I sparsely sample it, following standard practice in the literature. The sparse sampling steps are important and relatively complex in execution. Sparse sampling, the process of constructing lower frequency returns from higher frequency data, means for my work the process of constructing 1-minute samples using the TAQ tick data as the starting point (recall that my rationale for 1-minute sampling is explained above in the methodology). 
$\\[0.1in]$

I organize the data into minutes based upon Open, High, Low, Close (OHLC) quote prices in turn based on the NBBO. Each Bid or Ask quote price thus refers to a NBBO price. Each Exchange publishes its top-of-book price and size at the NBBO price and the NBBO is changed when an Exchange publishes a quote that improves the NBBO. Bid/Ask quotes persist until there is a change due to a trade, improved price, canceled Bid or Ask. I carry forward the NBBO Bid and Ask from the last NBBO event until there is a new NBBO event, which I detect by the presence of a NB flag on the event entry. When the market is crossed (i.e. NBBO Bid > NBBO Ask ) it is not possible to know what is the correct price, so I use the last good NBBO Bid and Ask, including situations where Bid == Ask. OpenMinTime corresponds to OpenBidPrice and OpenAskPrice. CloseMinTime corresponds with CloseBidPrice and CloseBidAsk, which, for my 1 minute data is HH:MM:59.999999999. If there are no changes to both the Bid and Ask in the NBBO during the 1-minute period then, for the Bid: OpenMinTime $=$ HighBidTime $=$ LowBidTime; and OpenBidPrice $=$ HighBidPrice $=$ LowBidPrice $=$ CloseBidPrice. Whilst for the Ask: OpenMinTime $=$ HighAskTime $=$ LowAskTime; and OpenAskPrice $=$ HighAskPrice $=$ LowAskPrice $=$ CloseAskPrice.
$\\[0.1in]$

In order to negate or reduce the impact of microstructure noise, and in particular Bid-Ask bounce, I compute a weighted mid-price (WMP) using the CloseBid and CloseAsk prices and sizes:
$\\[0.1in]$

\begin{align}
    \mathrm{WMP} = I P^a + \left( 1 - I \right) P^b
\end{align}

$\\[0.1in]$
where the weight I is given by the imbalance:
$\\[0.1in]$

\begin{align}
    I = \frac {Q^b} {Q^b + Q^a}
\end{align}

$\\[0.1in]$
where $Q^b$ is the bid size (that is, total volume at the best bid) and $Q^a$ is the ask size. I enunciate microstructure noise, Bid-Ask bounce, WMP, and my motivation for using WMP for this research, in detail below (see section entitled 'Model').
$\\[0.1in]$

## Empirical analysis

In this section I show within my AAPL, JPM, and EURUSD 1-minute returns data some of the key features of observed HF returns distributions that motivate the study of financial time-series volatility. For example, how persistent is volatility? And how quickly do financial markets forget large volatility shocks? Figures 7 and 8 show 1-minute WMP returns for AAPL, JPM, and EURUSD exhibit significant autocorrelations over many lags. Initially, volatility models did a relatively poor job of capturing this phenomena. In any stationary ARCH or GARCH model, memory decays exponentially fast. For example, if $\{ \epsilon_t \}$ are ARCH($1$), the $\{ \epsilon_t^2 \}$ have autocorrelations $\rho_k = \alpha^k$. Specifically, if $\alpha = 0.8$ and $k = 20$, then $\rho_{20} = 0.012$. This is likely to be an unrealistically fast decay. Conversely however, for any integrated ARCH or GARCH, $\rho_k = 1$ for all $k$, so there is no decay at all. This of course is unrealistically slow. Neither reflects what my data shows. At one level of abstraction, the progression from ARCH(1) to ARCH($q$) to GARCH represented an attempt to allow for the strong volatility persistence observed in actual data. Typically the exponential decay inherent in any stationary ARCH or GARCH model is too rapid to adequately describe the data, especially HF data, forcing the estimated models to be integrated. In reality, however, volatility may not be integrated, and the behavior of the estimated ARCH and GARCH models may simply be a signal that the memory is decaying relatively slowly compared to the exponential rate, or at several different rates.
$\\[0.1in]$

In section 5.10 Relative Performance Assessment, below, I argue for using the Heterogeneous Autoregressive (HAR) model of Corsi (2009) as a benchmark for my work. The HAR framework posits the conditional variance of discretely sampled returns as a linear function of the lagged squared return over the identical return horizon in combination with the squared returns over longer and/or shorter return horizons. This framework is obviously relevant to HF data settings and I review the broadly robust supporting literature in section 5.10. Motivated by the success of HAR-type models, a body of work following Corsi (2009) extends HAR by generalizing with jumps, leverage effects, and other nonlinear behaviors. For example, Bollerslev, Litvinova & Tauchen (2006) emphasized the importance of asymmetric leverage effects when modeling the volatility process and Corsi, Pirino &  Reno (2010) included jumps. Nonlinear effects, including structural breaks and regime-switches, were modeled by Scharth & Medeiros (2009) and McAleer & Medeiros (2008). Patton & Sheppard (2015) modeled equity price volatility with signed realized semi-variances, expositing the HAR-RS model, and showing that future volatility is more closely related to the volatility of past negative returns. Such results resonate more readily with my empirical analysis of AAPL, JPM, and EURUSD HF data below. But it is necessary to extend these paradigms further to better explain what is actually seen in the data, which I do in this study via use of a multifractal framework. Multifractality reflects the presence of both long-range dependence and hierarchical organization at many scales; the multifractal properties result from the explicit mechanism that future values of the process are nonlinearly influenced by the whole past history. This long-memory feature of the process is induced by regime switching of heterogeneous persistence: the process is driven by $\bar k$ independent Markov-switching processes with different, though tightly parametrized, transition probabilities.
$\\[0.1in]$

Figures 7 and 8 plot sample autocorrelograms for log returns and squared log returns respectively for the HF datasets considered in the Q-Q plots of figure 6. The autocorrelograms (which may also be referred to in the literature as autocorrelation function (ACF) plots or autocorrelation plots), enable visualization of the autocorrelation of the HF time-series at defined time lags. Where no relationship exists between the times-series and its lagged values, autocorrelations should be near zero for any and all time-lag separations. If however relationships exist, then one or more of the autocorrelations will be significantly non-zero. The autocorrelogram represents this visually.
$\\[0.1in]$

Figure 7 shows log return autocorrelograms for AAPL, JPM, EURUSD 1-min Bid, Ask, and Weighted Mid-Prices with 798, 593, and 1354 lags, which equate to approximately 1 day, respectively. 5% significance bands are shown as darker grey zones and the autocorrelograms are based upon full datasets (i.e. 27 months of HF data). Lags are plotted on the x-axis, versus correlations on the y-axis. In the case of AAPL, the top 3 charts in the panel, 1 day's lags equates to 798 lags, for JPM it is 593 lags, and for EURUSD it is 1,354 lags. 95% confidence intervals are also shown, where I have computed the standard deviation using Bartlett's (1946) formula; the upper and lower 5% significance limits for the autocorrelations are shown by the dark grey zones. I exclude the 0-lag autocorrelation for the purposes of drawing the plots because it is visually distorting. The autocorrelograms are constructed by Python scripts, which I wrote as follows:
$\\[0.1in]$

Let $\hat \mu$ estimate the mean, $\mu$, of one of my HF time series, $X_t$, where:
$\\[0.1in]$

\begin{equation}
    \hat \mu =
    \begin{cases}
    \mu  & \text{for  } \mu \text{  known} \\
    \frac {1} {n} \sum^n_{t=1} X_t & \text{for  } \mu \text{  unknown} 
    \end{cases}
\end{equation}

$\\[0.1in]$
I estimate the autocovariance function $\sigma \left( k \right)$ thus:
$\\[0.1in]$

\begin{align}
    \hat \sigma \left( k \right) = \frac {1} {n} \sum^n_{t=1} \left( X_t - \hat \mu \right) \left(X_{t+k} - \hat \mu \right), \; \; \; \; \; \; k = 0, 1, 2, ... K
\end{align}

$\\[0.1in]$
where $K = \mathrm{max lag}$. I estimate the autocorrelation function $\rho \left( k \right)$ thus:
$\\[0.1in]$

\begin{align}
    \hat \rho \left( k \right) = \frac {\hat \sigma \left( k \right)} {\hat \sigma \left( 0 \right)}, \; \; \; \; \; \; k = 0, 1, 2, ... K
\end{align}

$\\[0.1in]$
and note that $\hat \sigma \left( 0 \right)$ is an estimate of the sample variance, whilst $\hat \rho \left( 0 \right) \equiv 1$, by definition. 
$\\[0.1in]$

As noted above, I write onto the plot a confidence interval (95%) by computing the standard deviation using Bartlett's (1946) formula, which is based upon a general asymptotic expression for the variance of the autocorrelation coefficient. Theoretically, the formula upon which my code is based is written:
$\\[0.1in]$

\begin{align}
    \mathrm{Var} \{\hat \rho \left( k \right) \} = \frac {1} {n} \sum_{i=-\infty}^{\infty} \left[ \rho^2 \left(i \right) + \rho \left( i - k \right) \rho \left(i + k \right) - 4 \rho \left( i \right) \rho \left( k \right) \rho \left( i - k \right) + 2 \rho^2 \left( i \right) \rho^2 \left( k \right) \right]
\end{align}

$\\[0.1in]$
where $\hat \rho \left( k \right)$ assumes $\mu$ is unknown. These confidence intervals may also be referred to as Bartlett bands, in this case 95% Bartlett bands, as in, for example Andersen, Bollerslev, Christoffersen & Diebold (2011).
$\\[0.1in]$

In my actual Python implementation, the autocorrelations $\rho \left( k \right)$ are replaced by their estimates $\hat \rho \left( k \right)$ for $\left| k \right| \leq K$, and the limits of summation are in fact bounded by the assumption that $\rho \left( k \right) = 0$ for all $k$ so that $\left| k \right| > K$.
$\\[0.1in]$

It is widely shown in the literature that returns are strongly serially correlated, often with very slowly decaying autocorrelations and that volatility long memory dynamics are illustrated by hyperbolic decay (Bollerslev (2010) provides a summary of the relevant literature). Predictability may show up as significant autocorrelations in returns, whilst volatility
clustering will show up as significant autocorrelations in squared returns. In figure 7 all autocorrelograms exclude the 0-lag autocorrelation, being typically 1 as the data is correlated with itself, in order to scale the y-axis better for lags $k = 1, 2, ... K$. In all cases therefore the first lag is negative and I observe a significant and initially sharply decaying autocorrelation beyond the first lag. Thereafter the autocorrelations are dispersed, many are significant, some not, and exhibit a channel-type pattern, quite tightly in the case of JPM and more loosely in the cases of AAPL and EURUSD. In the 1-day autocorrelogram panel there does not appear to be a visually obvious difference between the WMP and the Bid or Ask plots.

I compute Ljung-Box (1978) Q-test (LBQ) statistics to test the significance of the autocorrelations. Table 8 reports LBQ statistics and associated p-values for AAPL, JPM, and EURUSD HF Bid, Ask, and Weighted Mid Price 1-minute log returns, respectively. The LBQ is a modified version of the Box-Pierce chi-square statistic, the latter being a simplified version of the former.  Similar assessment may also be conducted using a Breusch-Godfrey (1978) test or Durbin-Watson (1950, 1951) test. Instead of testing randomness at each distinct lag, it tests the autocorrelation based on a given number of lags $k = 0, 1, 2, ..., K$. I test the null hypothesis that my HF log returns are independently distributed and correlations are zero, thus:
$\\[0.1in]$

\begin{align}
    H_0: \rho \left( 1 \right) = 0, ..., \rho \left( k \right) = 0
\end{align}

$\\[0.1in]$
I compute the LBQ statistic thus:
$\\[0.1in]$

\begin{align}
    Q = n \left( n+2 \right) \sum^h_{k=1} \frac {\hat \rho^2_k} {\left( n - k \right)}
\end{align}

$\\[0.1in]$
where $n$ is the number of observations in my 1-minute log return datasets for AAPL, JPM, and EURUSD respectively, $\hat \rho_k$ is the $k$'th 1-minute log return autocorrelation, and $h$ is the number of lags tested, which is 10 for the results shown in table 8. Under $H_0$, $Q$ asymptotically follows a $\chi^2_{\left( h \right)}$. For significance level $\alpha$, the critical region for rejection of the null hypothesis of randomness is $Q > \chi^2_{1-\alpha, h}$, where $\chi^2_{1-\alpha, h}$ is the $1-\alpha$'th quantile of the chi-squared distribution with $h$ degrees of freedom.
$\\[0.1in]$

# Model

## Mitigating microstructure noise

I use a WMP to mitigate microstructure noise in this research and my primary justification for use of the WMP is empirical, not theoretical. Trading via limit orders and market orders has become the dominant form of trading in most exchanges around the world, whether these are pure electronic limit order markets, or hybrid markets in which limit order traders are in competition with floor traders, specialists, or dealers (see for example Roşu (2019, 2009)). The order book matches orders on a price-time priority and keeps track of all previously submitted buy and sell limit orders. Though all orders are anonymous, the econometrician (along with the much broader universe of market participants) may see the orders at each price level, and thus have information about the instantaneous supply and demand in the market. With the benefit of this information, she might quite naturally choose to build a mid-price, thus:
$\\[0.1in]$

\begin{align}
    M = \frac {1} {2} \left( P^a + P^b \right)
\end{align}

$\\[0.1in]$
where $P^b$ is the best (that is, the highest) Bid price and $P^a$ is the best (lowest) ask price. This measure is broadly accepted as a notion of 'fair' price by academics and practitioners.
$\\[0.1in]$

An important effect of the transparency of the limit order book is that the volume at the best Bid and Ask prices provides a strong signal of the next price move. To improve upon the mid-price therefore, the econometrician may choose instead to follow a widely used method among practitioners, who also pay close attention to the weighted mid-price (WMP):
$\\[0.1in]$

\begin{align}
    \mathrm{WMP} = I P^a + \left( 1 - I \right) P^b
\end{align}

$\\[0.1in]$
where the weight I is given by the imbalance:
$\\[0.1in]$

\begin{align}
    I = \frac {Q^b} {Q^b + Q^a}
\end{align}

$\\[0.1in]$
where $Q^b$ is the bid size (that is, total volume at the best bid) and $Q^a$ is the ask size.
$\\[0.1in]$

The volume at the best Bid and Ask prices has been shown to be a strong predictor of the next traded price. This feature of imbalances has been reported by, for example, Stoikov (2020), Dacorogna (2018), Lehalle & Mounjid (2018), Stoikov & Waeber (2016), Gould & Bonart (2015), Lehalle & Laruelle (2014), Lipton, Pesavento & Sotiropoulos (2013), Burlakov, Kamal & Salvadore (2012), Robert & Rosenbaum (2012), Avellaneda, Reed & Stoikov (2011), and Gatheral & Oomen (2010)).
$\\[0.1in]$

WMP is widely used in practice (see for example Stoikov (2020), Dacorogna (2018), Lehalle & Mounjid (2018), and Gatheral & Oomen (2010)), and has also been used by researchers (Ait-Sahalia, Mykland & Zhang (2010), for example, compute log-returns from quotes using a bid-ask midpoint, weighted by the respective depth of the two sides). The typical explanation (from a practitioner) for use of WMP is that, the larger quantity of shares on the Bid than on the Ask typically observed, indicates greater buying pressure, and therefore the "true" price is closer to the Ask than to the Bid. Thus, it follows, and is a reasonable assumption, that market buy/sell orders arriving are likely to be roughly matched and thus use the quantities at the Bid and Ask to adjust for the concomitant movement in the price. It is this mechanism that negates or reduces Bid-Ask bounce.
$\\[0.1in]$

## Markov-switching multifractal

Having prepared the HF $\mathrm{WMP}$ time series datasets, and cognizant of the suppression of microstructure noise thus achieved, I move on to construct the bivariate MSM of Calvet & Fisher (2004, 2008b). My $\mathrm{WMP}_t$ series, defined as the natural log of the weighted mid-price of the securities or exchange rate, follows the construction in equations 6.2 and 6.3. I define the innovations $x_t \equiv \mathrm{WMP}_t - \mathrm{WMP}_{t-1}$ and consider them within an economy that has $\bar k$ components $M_{1,t}, M_{2,t}, \: ..., \: M_{\bar k,t}$, which decay at heterogeneous frequencies $\gamma_1, \gamma_2, ..., \gamma_{\bar k}$. 
$\\[0.1in]$

I model the innovations $x_t \equiv \mathrm{WMP}_t - \mathrm{WMP}_{t-1}$ thus:
$\\[0.1in]$

\begin{equation}
    x_t = \sigma \left( M_{1,t}, M_{2,t}, ..., M_{\bar k,t} \right)^\frac {1}{2} \epsilon_t
\end{equation}

$\\[0.1in]$
The parameter $\sigma$ is a positive constant. The random variables $\epsilon_t$ are $\mathrm{i.i.d.}$ standard Guassian $\mathcal{N}(0,1)$. The random multipliers or volatility components $M_{k,t}$ are persistent, non-negative and satisfy $\mathbb{E} \left( M_{k,t} \right) = 1$. The multipliers $M_{1,t}, M_{2,t}, \: ..., \: M_{\bar k,t}$ at a given time $t$ may be considered to be statistically independent. The parameter $\sigma$ is thus equal to the unconditional standard deviation of the innovation $x_t$.
$\\[0.1in]$

Equation 6.4 defines a stochastic volatility model $x_t = \sigma_t \epsilon_t$ with the multiplicative structure $\sigma_t = \sigma \left( M_{1,t}, M_{2,t}, ..., M_{\bar k,t} \right)^\frac {1}{2}$. I stack the period $t$ volatility multipliers into a vector $M_t = \left( M_{1,t}, M_{2,t}, ..., M_{\bar k,t} \right)$. For any $m = \left( m_1, m_2, ..., m_{\bar k} \right) \in \mathbb{R}^{\bar k}$, let $g \left( m \right)$ denote the product $\prod_{i=1}^{\bar k} m_i$. Volatility at time $t$ is then $\sigma_t = \sigma {\left[ g \left( M_t \right) \right]}^\frac {1} {2}$.
$\\[0.1in]$

The properties of volatility are driven by the stochastic dynamics of the vector $M_t$. To simplify the set up Calvet & Fisher assume that $M_t$ is first-order Markov, which facilitates the construction through time of $\{ x_t \}$ and permits maximum likelihood estimation. $M_t$ is denoted as a volatility state vector, whilst each component $M_{k, t}$ is denoted as a state variable. The econometrician observes the returns $x_t = \sigma {\left[ g \left( M_t \right) \right]}^\frac {1} {2} \epsilon_t$, but not the vector $M_t$ itself. The vector $M_t$ is therefore latent, and must be inferred recursively by Bayesian updating. 
$\\[0.1in]$

Each $M_{k,t}$ follows a process that is identical except for time scale. Assume that volatility state variables have been constructed up to date $t - 1$. For each $k \in \{ 1, 2, ..., \bar k \}$, the next period multiplier $M_{k,t}$ is drawn from a fixed distribution $M$ with probability $\gamma_k$ and is otherwise equal to its current value: $M_{k,t} = M_{k, t-1}$. The dynamics of $M_{k,t}$ can therefore be summarized thus:
$\\[0.1in]$

\begin{equation}
    M_{k,t} =
    \begin{cases}
    \text{ drawn from distribution M }  & \text{ with probability } \gamma_k \\
    M_{k,t-1} & \text { with probability }  1 - \gamma_k
    \end{cases}
\end{equation}

$\\[0.1in]$
Equation 6.5 is of course the discrete time analogue of equation 3.10; Calvet & Fisher assume for simplicity that switches and new draws from $M$ are $\mathrm{i.i.d.}$ across $k$ and $t$. The volatility components $M_{k,t}$ thus differ in their transition probabilities $\gamma_k$ but not in their marginal distribution of $M$.
$\\[0.1in]$

The transition probabilities $\gamma \equiv \left( \gamma_1, \gamma_2, ..., \gamma_{\bar k} \right)$ may be specified thus: 
$\\[0.1in]$

\begin{align}
    \gamma_k = 1 - \left( 1 - \gamma_1 \right)^{b^{k - 1}}
\end{align}

$\\[0.1in]$
where $\gamma_1 \in \left( 0, 1 \right)$ and $b \in \left( 1, \infty \right)$. Since $1 - \gamma_k = {\left( 1 - \gamma_1 \right)}^{b^{k-1}}$, the logarithms of staying probabilities are exponentially decreasing with $k$. 
$\\[0.1in]$

The parameter $\gamma_{\bar k}$ thus denotes the high frequency switching probability, whilst $b$ denotes the frequency growth rate. In this setting, a process with very persistent components (equivalently a very small parameter $\gamma_1$), with small values of $k$, also has small $\gamma_1 b^{k-1}$, so that the transition probability satisfies $\gamma_k \sim \gamma_1 b^{k-1}$. At higher frequencies $\left( \gamma_k \sim 1 \right)$, the rate of increase slows down and condition 6.6 guarantees that the parameter $\gamma_k$ remains lower than 1. In empirical settings, it is numerically convenient to estimate parameters with the same order of magnitude. Since $\gamma_1 < \; ... \; < \gamma_{\bar k} < 1 < b$, the set of transition probabilities is specified by $\left( \gamma_{\bar k}, b \right)$.
$\\[0.1in]$

The parameter $\bar k$ determines the number of volatility frequencies in the model. It is fixed, and thus a model selection problem. The multifractal construction imposes only minimal restrictions on the marginal distribution of the multipliers: $M \geq 0$ and $\mathbb{E} \left( M \right) = 1$. Calvet & Fisher show that $M$ is drawn from a family of unit mean distributions that are specified by a single parameter $m_0$. I use the simple binomial set up for this research, where binomial random variables take the values $m_0$ or $2 - m_0$ with equal probability.
$\\[0.1in]$

The full parameter vector is thus:
$\\[0.1in]$

\begin{equation}
    \psi \equiv \left( m_0, \sigma, b, \gamma_{\bar k} \right) \in \mathbb{R}^4
\end{equation}

$\\[0.1in]$
where $m_0$ characterizes the distribution of the multipliers, $\sigma$ is the unconditional standard deviation of $\mathrm{WMP}$ returns, and $b$ and $\gamma_{\bar k}$ define the set of switching probabilities. 
$\\[0.1in]$

## Maximum likelihood estimation

I have shown that, in my set up, the multiplier $M$ has a discrete distribution, therefore finite number of volatility states. Standard filtering methods thus provide the likelihood function in closed form. Let the multiplier $M$ take a finite number of values $b_m$, so that the vector $M_t = \left( M_{1, t}, M_{2, t}, ..., M_{{\bar k}, t} \right)$ may therefore take $d = b^{\bar k}_m$ values, $m^1, m^2, ..., m^d \in \mathbb{R}^{\bar k}$.

The dynamics of the Markov chain $M_t$ are characterized by a transition matrix $A = {\left( a_{i,j} \right)}_{1 \leq {i, j} \leq d}$, with components $a_{i, j} = \mathbb{P} \left(M_{t+1} = m^j \mid M_t = m^i \right)$. I construct these components thus:
$\\[0.1in]$

\begin{equation}
    a_{i, j} = \prod^{\bar k}_{k = 1} \left[ \left( 1 - \gamma_k \right) 1_\{m^i_k = m^j_k\} + \gamma_k \mathbb{P} \left( M = m^j_k \right) \right]
\end{equation}

$\\[0.1in]$
where $m^i_k$ denotes the $m$th component of vector $m^i$, and $1_\{m^i_k = m^j_k\}$ is a dummy variable equal to 1 if $m^i_k = m^j_k$ and 0 otherwise.
$\\[0.1in]$

The density of the innovation $x_t$ conditional on $M_t$ is:
$\\[0.1in]$

\begin{equation}
    f_{x_t} \left( x \mid M_t = m^i \right) = {\left[ \sigma g \left( m^i \right) \right]}^-1 n \left[ \frac {x} {\sigma g \left( m^i \right)} \right]
\end{equation}

$\\[0.1in]$
where $n \left(.\right)$ denotes the density of a standard normal. 
$\\[0.1in]$

The conditional probablities $\prod^j_t \equiv \mathbb{P} \left( M_t = m^j \mid x_1, ..., x_t \right)$ over the unobserved states $m^1, m^2, ..., m^d$ can be stacked in the row vector $\prod_t = \left( \prod^1_t, \prod^2_t, ..., \prod^d_t \right) \in \mathbb{R}^d_+$. Letting $\iota = \left( 1, ..., 1 \right) \in \mathbb{R}^d$, I know that $\prod_t i\prime = 1$.
$\\[0.1in]$

I use Bayes' rule to update $\prod_{t+1}$ (the new belief) from $\prod_t$ (the old belief) and the new innovation $x_{t+1}$, using the function: 
$\\[0.1in]$

\begin{equation}
    \omega \left( x_t \right) = \left[ \frac {n \left( \frac {x_t} {\sigma g \left( m^1 \right)} \right)} {\sigma g \left( m^1 \right)}, \frac {n \left( \frac {x_t} {\sigma g \left( m^2 \right)} \right)} {\sigma g \left( m^2 \right)}, ..., \frac {n \left( \frac {x_t} {\sigma g \left( m^d \right)} \right)} {\sigma g \left( m^d \right)} \right]
\end{equation}

$\\[0.1in]$
From 6.10 I can express the conditional probability $\prod_{t+1}$ as a function of the observation $x_{t+1}$ and the probability $\prod_t$ calculated in period $t$, thus: 
$\\[0.1in]$

\begin{equation}
    \Pi_{t+1} = \frac {\omega \left( x_{t+1} \right) \ast \left( \prod_t A \right)} {\left[ \omega \left( x_{t+1} \right) \ast \left( \prod_t A \right) \right] \iota\prime}
\end{equation}

$\\[0.1in]$
where $x \ast y$ denotes the Hadamard product $\left( x_1 y_1, x_2 y_2, ..., x_d y_d \right)$ of any $x, y \in \mathbb{R}^d$. With this formulation $\prod_t$ may be recursively calculated. Since the multipliers $\left( M_{1, 1}, M_{2, 1}, ..., M_{{\bar k}, 1} \right)$ are independent, the initial vector $\prod_0$ is uniquely determined by $\prod^j_0 = \prod^{\bar k}_{l = 1} \mathbb{P} \left(M = m^j_l \right)$ for all $j$.
$\\[0.1in]$

## More volatility states

The dimension of the transition matrix in the binomial setting is $2 \bar k \ast 2 \bar k$.  Thus the computational load arising from this operation increases exponentially (the curse of dimensionality). Additionally, Bayesian updating and estimation involve a recursion that increases the computational load in proportion with the number of observations in the dataset.  The UHT data is 372,607 returns; it is thus necessary for the model to incorporate several features that enable it to handle high $\bar k$ settings and large HF datasets.
$\\[0.1in]$

A disadvantage of the exact maximum likelihood estimator is that it becomes computationally demanding for $\bar k \geq 10$, since the dimension of the transition matrix grows at a rate of $2^{\bar k}$. One sees easily how the challenge scales: Under the binomial specification, at $\bar k = 10$ MSM has 1,024 volatility states, whilst at $\bar k = 11$ it has 2,048 volatility states. The transition probability matrix $A$ (equation 6.8 is doubly stochastic and of dimension $2^{\bar k}$ by $2^{\bar k}$. It is necessary to write it for each step of the Bayesian recursion. In turn, the Bayesian recursion (equations 6.10 and 6.11 requires $T$ steps, where $T$ is the length of the time series. Higher $\bar k$ in conjunction with larger $T$ (high frequency data settings, for example) thus present a formidable computational challenge! 
$\\[0.1in]$

In their 2004 paper, Calvet & Fisher, introducing the MSM, observed "Such a model [the binomial MSM] could be very unwieldy as the number $\bar k$ becomes very large". Calvet, Fisher & Thompson (2006) develop a particle filter to accomodate larger state spaces. Several other papers in the multifractal literature acknowledge the computational challenge of higher $\bar k$ and restrict their work accordingly: $\bar k \leq 3$ (Idier (2009)); $\bar k \leq 5$ (Calvet, Fisher & Thomson (2006) when numerically deriving maximum likelihood estimates); $\bar k \leq 7$ (Chen, Diebold & Schorfheide (2013), Hainaut (2012)); $\bar k \leq 8$ (Zikes, Barunik & Shenai (2018), (Alausa (2014), Idier (2008)); and $\bar k \leq 10$ (Wang, Wu & Yang (2016), Calvet & Fisher (2004)). 
$\\[0.1in]$

The exact maximum likelihood estimator is also not applicable to the log-normal MSM process, where the state space of the hidden Markov chain is infinite. These issues motivated Lux (2008) to develop a generalized method of moments (GMM) approach, which has been shown by Lux (2008, 2018), Liu, Matteo & Lux (2007), Lux, Morales-Arias & Sattarhoff (2011), and Segnon, Lux & Gupta (2015) to work for a wide range of MSM specifications and requires only moderate computational resources. Similarly, Bacry, Kozhemyak & Muzy (2008, 2012) use a GMM approach to estimate parameters of the Multifractal Random Walk (MRW). A potential drawback to the GMM estimator of Lux (2008) is that it is applied to the first differences rather than levels of the process and this makes the identification of the transition probability parameters $b$ and $\gamma_{\bar k}$, even when the sample size is very large. Lux (2008, 2018), Liu, Matteo & Lux (2007), Lux, Morales-Arias & Sattarhoff (2011), and Segnon, Lux & Gupta (2015) circumvent this circumvent this problem by setting $b$ and $\gamma_{\bar k}$ arbitrarily, then estimate $m_0$ and $\hat \sigma$. This may however be quite restrictive.
$\\[0.1in]$

## The likelihood function

I estimate the model by maximum likelihood. Let $L = \left( x_1, x_2, ..., x_T; \psi \right) = \sum^T_{t=1} \mathrm{ln} f \left(x_t \mid x_1, x_2, ..., x_{t - 1} \right)$. Bayes' rule implies that:
$\\[0.1in]$

\begin{align}
    f \left(x_t \mid x_1, x_2, ..., x_{t - 1} \right) &= \sum^d_{i=1} \mathbb{P} \left( M_t = m^i \mid x_1, x_2, ..., x_{t - 1} \right) f \left(x_t \mid M_t = m^i \right) \\
    &= \sum^d_{i=1} \mathbb{P} \left( M_t = m^i \mid x_1, x_2, ..., x_{t - 1} \right) \frac {1} {\sigma g \left( m^i \right)} n \left( \frac {x} {\sigma g \left( m^i \right)} \right)
\end{align}

$\\[0.1in]$
Providing the log likelihood function:
$\\[0.1in]$

\begin{equation}
    \mathrm{ln} L \left( x_1, x_2, ..., x_T; \psi \right) = \sum_{t=1}^T \ln \left[ \omega \left( x_t \right) \cdot \left( \Pi_{t-1} A \right) \right]
\end{equation}

$\\[0.1in]$
Calvet & Fisher (2004, 2008b) show that the maximum likelihood estimator is consistent and asymptotically efficient as $T \rightarrow \infty$.
$\\[0.1in]$

## Basin-hopping

Nonlinear, non-Gaussian state space models such as MSM do not generally allow for an analytical representation of their likelihood function. In the literature, sequential Monte Carlo or particle filtering methods has been used as an alternative for parameter estimation. Calvet, Fisher & Thompson (2006) introduce a particle filter and Liu & Lux (2015) apply the Nelder-Mead or simplex method, for example. Lux (2018) studies the performance of two numerical optimization methods, the Iterated Filtering I and II algorithms of Ionides, Bhadra, Atchade & King (2015) and Ionides, Breto & King (2006) respectively, versus a baseline Nelder-Mead approach. I observe that in implementation, the Iterated Filtering I algorithm appears to be close to a repeated application of the perhaps better known Liu-West filter (Liu and West, 2001), whilst, again in implementation, the Integrated Filtering II algorithm  appears close to an iterated version of the bootstrap filter proposed by Rios and Lopes (2013).
$\\[0.1in]$

I enable exact maximum likelihood with $\bar k \leq 13$. This brings to bear 8,192 volatility states, which is eight times higher than previous literature. In order to achieve this, it is necessary to greatly speed up the model, in order to handle the larger transition matrix $A$ and the also, in this study, a larger $T$ resulting from use of HF data. Firstly, I optimize the code and technical architecture (hardware and software). Secondly, I introduce a two-step basin-hopping method, following Wales & Doye (1997). The basin-hopping framework has been broadly used outside of finance for hard nonlinear optimization problems with multiple variables and modalities (for example, characterization of molecular states in statistical physics - see Olsen, et al. (2012) for a summary of non-financial applications).
$\\[0.1in]$

Basin-hopping is a stochastic algorithm that combines heuristic procedures with local searches to perform an enhanced exploration of the given variable space, conducted as a series of perturbations, followed by local optimization. The algorithm is iterative with each cycle comprising a random perturbation of the binomial MSM parameter vector $\theta = \left( b, \: m_0, \: \: \gamma_{\bar k}, \: \bar \sigma \right)$, local minimization, and acceptance or rejection of the new parameter vector based upon the minimized function value. The acceptance test used in my implementation of the algorithm is the Metropolis criterion, a variant of the stochastic Monte Carlo procedure. My implementation of the algorithm is illustrated below:
$\\[0.1in]$

*Algorithm 1: Basin-hopping*
$\\[0.1in]$

Initialize the parameter vector $\: \theta_0 = \left( b, \: m_0, \: \: \gamma_{\bar k}, \: \bar \sigma \right)$.  

Set $\: b = 1.001$, $m_0 = 1$, $\gamma_{\bar k} = 1e-3$, $\bar \sigma = 1e-4$,  

Where bounds $= ((1.001,50),(1,1.999),(1e-3,0.999),(1e-4,5))$,

While $\theta_t$ not converged do:

$i \longleftarrow 0$,

$\theta_i \longleftarrow$  Random initial vector in variable space,

$LL_i \longleftarrow$ Run local minimization function once for each basin-hopping step,

While stop not satisfied do:

$\theta_{i+1} \longleftarrow$ Perturb $\left( \theta_i \right)$,

$LL_{i+1} \longleftarrow$ Local minimization function $\mid$ args = minimizer_kwargs, $\theta_{i+1}$,  

If $f \left( LL_{i+1} \right ) < f \left( LL_i \right )$ then, 

$i \longleftarrow i+1$, 

End.

Result: Final parameter vector $\: \theta^* = \left( b, \: m_0, \: \: \gamma_{\bar k}, \: \bar \sigma \right)$, vector of log likelihoods at global minimum.  <br>
$\\[0.1in]$

In a stochastic global optimization setting the econometrician is unable to observe if a true global minimum has been found. I therefore run the algorithm from a number of different random starting points to ensure that the lowest minimum found in each case has converged to the global minimum. I use a bounded minimization to estimate the starting points and provide an interval constraint for $b$ and $\bar \gamma_k$, whilst optimizing for $m_0$ within each bound. I set the bounds equal to 1.5, 5, 15, 30 for $b$, and 0.1, 0.5, 0.9, 0.95 for $\bar \gamma_k$ respectively.  An optimal $m_0$ is thus computed with likelihood for each set of bounds, resulting in a grid of starting values for the basin-hopping algorithm plus associated vector of log-likelihood values.
$\\[0.1in]$

## Out-of-sample performance evaluation

Out-of-sample performance is assessed via volatility forecasts. I conduct typical full-sample Mincer & Zarnowitz (1969) forecast rationality tests, following the approach commonly used in the literature (see Diebold (2017) and Patton & Sheppard (2009) for extensive reviews of the method). 
$\\[0.1in]$

Writing the target value to forecast at time $t$ using information up to time $t-h$ as $y_t$ and denoting the forecast by $y_{t \mid t-h}$, the Mincer & Zarnowitz (MZ) regression may be written thus:
$\\[0.1in]$

\begin{equation}
    x^2_{t+h} = \gamma_0 + \gamma_1 \mathbb{E} x^2_{t+h} + u_{t, h}
\end{equation}

$\\[0.1in]$
where $h$ is the forecast horizon and $u_{t,h}$ is the residual. If the forecasts are unbiased, the constant $\gamma_0$ should be statistically insignificantly different from zero; if the forecasts are optimal, the slope $\gamma_1$ should be statistically insignificantly different from unity. The $\mathrm{MSE}$ quantifies forecast errors, thus: 
$\\[0.1in]$

\begin{equation}
    \mathrm{MSE} = L^{-1} \sum^T_{t=T-L+1} {\left( x^2_t - \mathbb{E}_{t-1} x^2_t \right)}^2
\end{equation}

$\\[0.1in]$
And the coefficient of determination is defined thus:
$\\[0.1in]$

\begin{equation}
    R^2 = 1 - \frac {\mathrm{MSE}} {L^{-1} \sum^T_{t=T-L+1} {\left( x^2_t - \sum^T_{i=T-L+1}  x^2_i / L \right)}^2} 
\end{equation}

$\\[0.1in]$
My null hypothesis therefore is that $\gamma_0 = 0$ and $\gamma_1 = 1$, jointly:
$\\[0.1in]$

\begin{equation}
    H_0 \: \: : \: \: \gamma_0 = 0 \cap \gamma_1 = 1
\end{equation}

$\\[0.1in]$
For a good forecast, therefore, $b_0 \approx 0$, $b_1 \approx 1$, $\mathrm{MSE}$ is low, and $R^2$ is high. 
$\\[0.1in]$

With $F_{t-1} = \left \{ r_1, \: ..., \: r_{t-1} \right \}$ I have:
$\\[0.1in]$

\begin{equation}
    E \left[ \sigma_t^2 \mid F_{t-1} \right] = \bar \sigma^2 \sum_{i, j=1}^d m^j \Pi_{t-1}^i a_{ij}
\end{equation}

$\\[0.1in]$
Since:
$\\[0.1in]$

\begin{equation}
    E \left[ M_t \mid F_{t-1} \right] = \sum_{j=1}^d m^j P \left( M_t = m^j \mid F_{t-1} \right)
\end{equation}

$\\[0.1in]$
Where:
$\\[0.1in]$

\begin{equation}
    P \left( M_t = m^j \mid F_{t-1} \right) = \sum_{i=1}^d P \left( M_t = m^j, M_{t-1} = m^i \mid F_{t-1} \right)
\end{equation}

$\\[0.1in]$
And:
$\\[0.1in]$

\begin{equation}
    P \left(M_t = m^j, M_{t-1} = m^i \mid F_{t-1} \right) = P \left(M_{t-1} = m^i \mid F_{t-1} \right) \: P \left( M_t = m^j \mid M_{t-1} = m^i \right) = \Pi_{t-1}^i a_{ij}
\end{equation}
$\\[0.1in]$

## Relative performance assessment

I compare out-of-sample forecasts from the MSM to the relative accuracy of out-of-sample forecasts produced by a RV measure for each of the HF time series (AAPL, JPM, EURUSD) plus the daily DEXJPUS time series. I use the heterogeneous autoregressive (HAR) forecasting model of Corsi (2009) for the comparison. The HAR model is popular in practice because it captures long-memory-type properties of quadratic variation, while being simpler to estimate than fractionally integrated processes. The HAR model has been shown to consistently provide strong out-of-sample volatility forecasts, including recent studies by Clements & Preve (2019), and Bollerslev, Patton & Quaedvlieg (2018, 2016). In a seminal study, Liu, Patton & Sheppard (2015) used HAR to compare out-of-sample performance across more than 420 realized measures and 'jump-robust' estimators. In earlier literature, Corsi, Mittnik, Pigorsch & Pigorsch (2008) used HAR to analyze the volatility of realized volatility;  Bollerslev, Tauchen & Zhou (2008) used HAR to analyze risk-return tradeoffs; Clements, Galvao & Kim (2008) implemented HAR for risk management with VaR measures and Andersen, Bollerslev & Diebold (2007) used an extension of HAR to forecast the volatility of stock prices, foreign exchange rates and bond prices. In all these studies HAR generally performed strongly.
$\\[0.1in]$

Interestingly, the HAR model does not appear to have been used as a benchmark previously in the multifractal literature. If we allow that the econometrician's choice of benchmark is driven at least in part by her problem setting, then multifractal studies may be considered to sit broadly within 10 categories. I describe these categories and assign the literature to them as follows. They are, firstly: expository papers and papers that test for multifractality in data and confirm multiscaling (Liu, Matteo & Lux (2007), Saichev & Sornette (2006), Calvet & Fisher (2002), and Mandelbrot, Calvet & Fisher (1997c)); plus papers that examine local properties of multifractal processes and multifractal spectra (Mandelbrot, Calvet & Fisher (1997a, 1997b)). These papers by their nature have no need for benchmarks. 

Secondly, papers that introduce extensions of the MSM: Lux & Morales-Arias (2009) introduce the Markov-Switching Multifractal model of asset returns with Student-t innovations; Idier (2009) extends his earlier research with univariate and bivariate MSM (Idier (2008)) to forecast correlations with a novel MSM Dynamic Conditional Correlation (MSMDCC) model;  Calvet, Fisher & Thomson (2006) introduce a bivariate model of currency volatility, the bivariate MSM; and Calvet & Fisher (2001) extend the original MMAR of Mandelbrot, Calvet and Fisher (1997a, 1997b, 1997c) to the Continuous time Poisson multifractal. These papers tend to benchmark versus the models that they extend, so that, for example, Idier (2009) benchmarks with the binomial MSM (Calvet & Fisher (2004)) and the Dynamic Conditional Correlation model (Engle & Sheppard (2002)). Calvet, Fisher & Thomson benchmark to the CC-GARCH (Bollerslev (1990).
$\\[0.1in]$

Thirdly, research that provides alternative estimation and optimization methods. Lux (2008) proposes an alternative GMM estimator, thus enabling considerably less computationally demanding linear forecasting. In a subsequent paper (Lux (2018)) studies the performance of two numerical optimization methods, the Iterated Filtering I and II algorithms of Ionides, et al. (2006, 2015). In both these papers Lux's benchmarks are GARCH (Bollerslev (1986)), and FI-GARCH (Baillie, Bollerslev & Mikkelsen (1996)).
$\\[0.1in]$

Fourthly, work that assesses the performance of multifractal models, usually via volatility forecasts; these papers include Lux & Kaizoji (2007), Calvet & Fisher (2004), and Calvet & Fisher (2001). Lux, Morales-Arias & Sattarhoff (2011) forecast realized volatility using best linear forecasts derived from GMM estimation. These papers compare in- and out- of sample performance versus GARCH (Bollerslev (1986)), FI-GARCH (Baillie, Bollerslev & Mikkelsen (1996)), and MS-GARCH (Klaassen (2002)); whilst Lux, Morales-Arias & Sattarhoff benchmark their best linear forecasts versus realized volatility forecasts from RV-ARFIMA, and FI-GARCH (Baillie, Bollerslev & Mikkelsen (1996)).
$\\[0.1in]$

Fifthly, studies of volatility surfaces, smiles and smirks, and the use of multifractal models for option pricing. These papers include Calvet, Fearnley, Fisher, & Leippold (2015) and Calvet & Fisher (2001). Calvet, Fearnley, Fisher, & Leippold (2015) use GARCH (Bollerslev (1986)), ZGARCH (Zakoian (1994)), and the GJR-GARCH of Glosten, et al. (1993) for comparison results.
$\\[0.1in]$

Sixthly, studies that embed MSM within a consumption-based asset pricing equilibrium, being Calvet & Fisher (2007, 2008). 
$\\[0.1in]$

Seventhly, alternate multifractal formulations, such as the Multifractal Random Walk (MRW) of Bacry, Delour & Muzy (2001), also studied by Bacry, Kozhemyak & Muzy (2008), and the Self-Excited Multifractal (SEMF) of Filimonov & Sornette (2010). These papers are largely theoretical, employing data simulation to demonstrate key concepts. Bacry, Kozhemyak & Muzy (2008) empirically assess MRW performance using volatility forecasting and VaR forecasts. The benchmark for this work is GARCH (Bollerslev (1986)).
$\\[0.1in]$

Eighthly, research that uses the MSM framework to study inter-trade durations (Chen, Diebold & Schorfheide (2013)), and subsequent work that extends upon the characterization of inter-trade durations (Aldrich, Heckenbach & Laughlin (2014)). The latter study benchmarks against the former.
$\\[0.1in]$

Ninethly, research that focuses upon commodities. These papers include Wang, Wu & Yang (2016), who forecast crude oil return volatility with a binomial MSM and Segnon, Lux & Gupta (2015) who forecast carbon dioxide (CO2) price volatility using a Lognormal MSM. Alausa (2014) extends the bivariate MSM with a Vector Error Correction term to hedge crude oil with a VEC-MSM.  These studies benchmark against GARCH (Bollerslev (1986)), IGARCH (Engle & Bollerslev (1986)), EGARCH (Nelson (1991)), GJR-GARCH (Glosten, Jagannathan & Runkle (1993)), MRS-GARCH (Cai (1994) and Hamilton & Susmel 1994)), and FI-GARCH (Baillie, Bollerslev & Mikkelsen (1996)).
$\\[0.1in]$

Finally, studies that use MSM as an alternative stochastic process, for example Hainaut (2012) replaces the constant volatility of the Brownian motion in the Hull-White model with a MSM process. GARCH (Bollerslev (1986)) is again a benchmark, along with the original Hull-White (1990) model.
$\\[0.1in]$

Organized within these broad categories, with the exception of Lux, Morales-Arias & Sattarhoff (2011), it can be seen that the multifractal literature to date does not appear to have directly evaluated the performance of a multifractal model with high frequency data and/or RV.  Where HF data has been used, notably by Lux, Morales-Arias & Sattarhoff (2011), it is in the context of incremental model extension within a more widely used setting, with the benchmarks accordingly aligned. Lux, Morales-Arias & Sattarhoff (2011), Lux & Kaizoji (2007), Lux (2008), Lux & Morales-Arias (2009)) all use similar settings for their studies: GMM estimation, best linear forecasts, and benchmarks from the (G)ARCH-class - i.e. GARCH (Bollerslev (1986)), IGARCH (Engle & Bollerslev (1986)), EGARCH (Nelson (1991)), GJR-GARCH (Glosten, Jagannathan & Runkle (1993)), and FI-GARCH (Baillie, Bollerslev & Mikkelsen (1996)). The study by Lux, Morales-Arias & Sattarhoff (2011) uses high frequency data as a datasource, but for the empirical work computes daily returns and realized volatilities using 30-minute samples, following the method of Andersen, Bollerslev, Diebold & Labys (2003) for preliminary data preparation. Zikes, Barunik & Shenai (2018) and Aldrich, Heckenbach & Laughlin (2014) use HF data to study Markov-switching Multifractal Duration models; these authors therefore use alternate duration models as benchmarks: the Long Memeory Short Duration model of Deo, Hsieh & Hurvich and the Autoregressive Conditional Duration model of Engle & Russell (1998) in the case of Zikes, Barunik & Shenai, whilst the Truncated-MSMD of Aldrich, Heckenbach and Laughlin benchmarks against the original MSMD of Chen, Diebold & Shorfheide (2013) upon which their work extends.
$\\[0.1in]$

I study multifractal volatility forecasting, with a much enlarged state space, using HF data, whilst specifically taking microstructure noise into account; I do so, in part, to assess the viability of that setup as an alternative to established RV measures. This naturally directs me to an established RV benchmark, which in turn, based upon the above analysis, would also seem to bridge an interesting gap in the literature.
$\\[0.1in]$

HAR's formulation is an extension of the Heterogeneous ARCH, or HARCH, of Muller, Dacorogna, Dave, Olsen, Pictet, & Weizsacker (1997). HARCH defines a set of partial volatilities, $\sigma_j$ measured at different frequencies thus:
$\\[0.1in]$

\begin{align}
    \sigma^2_{j,t} = \mu_j \sigma^2_{j,t-1} + \left( 1 - \mu_j \right) {\left( \sum^{k_j}_{i=1} r_{t-i} \right)}^2
\end{align}

$\\[0.1in]$
where $k_j = p^{j-2} + 1$ for $j > 1$ with $k_1 \equiv 1$, $\mu_j = e^{-2 / \left(k_{j+1}-k_j \right)}$, and $p$ may be chosen by the econometrician. The HARCH process may then be defined as a linear combination of these partial volatilities:
$\\[0.1in]$

\begin{align}
    \sigma^2_{t} = C_0 + \sum^n_{j-1} C_j \sigma^2_{j,t} \; \; \; \; \; \mathrm{with} \; \; \; \; \; r_t = \sigma_t \epsilon_t
\end{align}

$\\[0.1in]$
where $\epsilon_t$ is i.i.d. $\mathcal{N} \left( 0, 1 \right)$
$\\[0.1in]$

Corsi's (2009) original HAR model extends HARCH by specifying RV as a linear function of daily, weekly, and monthly realized variance components, expressed thus:
$\\[0.1in]$

\begin{align}
    \mathrm{RV}_t = \beta_0 + \beta_1 \mathrm{RV}^d_{t-1} + \beta_2 \mathrm{RV}^w_{t-1} + \beta_3 \mathrm{RV}^m_{t-1} + \epsilon_t
\end{align}

$\\[0.1in]$
where $\beta_j$ for $j =\{0, 1, 2, 3\}$ are unknown parameters to be esimated, $\mathrm{RV}_t$ is the realized variance of day $t$, and $\mathrm{RV}^d_{t-1} = \mathrm{RV}_{t-1}$ , $\mathrm{RV}_{t-1}^w = \frac {1} {5} \sum_{i=1}^5 \mathrm{RV}_{t-i}$ , $\mathrm{RV}_{t-1}^m = \frac {1} {22} \sum_{i=1}^22 \mathrm{RV}_{t-i}$ denote the daily, weekly and monthly lagged variance, respectively. 
$\\[0.1in]$

The lag structure for HAR was fixed by Corsi (2009) at $\{1, 5, 22\}$ to mirror daily, weekly, and monthly components of the volatility process. Whilst the suitability of the specification has been challenged, it has nevertheless proved robust. Craioveanu & Hillebrand (2012) used a parallel computing set up to investigate all possible combinations of lags up to a maximum lag of 250, reporting results that corroborate Corsi's $\{1, 5, 22\}$ configuration.
$\\[0.1in]$

To fix the concepts and notation with my empirical setting, consider a general jump-diffusion model for the log-price $p$ of an asset:
$\\[0.1in]$

\begin{align}
    d p \left( t \right) = \mu \left( t \right) dt + \sigma \left( t \right) d W \left( t \right) + \kappa \left( t \right) d N \left( t \right)
\end{align}

$\\[0.1in]$
where $\mu$ is the instantaneous drift, $\sigma$ is the (stochastic) volatility, $W$ is a standard Brownian motion, $\kappa$ is the jump size, and $N$ is a counting measure for the jumps. In the absence of jumps the third term on the right hand side above is zero. The quadratic variation of the log-price process over period $t+1$ may be defined thus:
$\\[0.1in]$

\begin{align}
    \mathrm{QV}_{t+1} &= \lim_{n\rightarrow\infty} \sum_{j=1}^n r^2_{t + j / n} \;, \\
    r_{t + j / n} &= p_{t + j / n}  - p_{t + \left( j -1 \right) / n}
\end{align}

$\\[0.1in]$
where my HF price series at $t+1$ is assumed to be observed $n$ times $\{ p_{t + 1 / n} , ... , p_{t + 1 - 1 / n} , p_{t + 1} \}$.
$\\[0.1in]$

My objective is to compare the out-of-sample forecasts from the MSM to the relative accuracy of out-of-sample forecasts produced by a RV measure for each of the HF time series (AAPL, JPM, EURUSD) plus the daily DEXJPUS time series. I therefore do this with emphasis upon comparisons with a simple RV estimator, which is the empirical analog of QV:
$\\[0.1in]$

\begin{align}
    \mathrm{RV}_{t+1} = \sum_{j=1}^n r^2_{t + j / n}
\end{align}

$\\[0.1in]$
I thus estimate the HAR model using the out-of-sample dataset (i.e. 40% of the original time series held back for out-of-sample testing) for each time series, AAPL, JPM, EURUSD and DEXJPUS respectively, thus:
$\\[0.1in]$

\begin{align}
    \tilde \theta_{t+h} = \beta_{0, j, h} + \beta_{1, j, h} M_{j, t} + \beta_{2, j, h} \frac {1} {5} \sum_{k=0}^4 M_{j, t-k} + \beta_{3, j, h} \frac {1} {22} \sum_{k=0}^21 M_{j, t-k} + \epsilon_{j, t}
\end{align}

$\\[0.1in]$
where $M_{j,t}$ is the RV measure for each of the HF time series plus DEXJPUS, and $\tilde \theta_{t+h}$ is the volatility proxy 6.27. I estimate this regression separately for each forecast horizon, $h$, and from those estimates I obtain $h$-day ahead volatility forecasts, which I them compare with the volatility proxy. I re-estimate the model for each $h$-day using a rolling window. I use Mincer-Zarnowitz regressions, $R^2$, and MSE to compare the out-of-sample forecasts thus produced by HAR to out-of-sample forecasts from the MSM.
$\\[0.1in]$

# Empirical study

I divided the HF datasets into in- and out-of-sample sub-sets, on a 60:40 basis. I thus used 60% of the data for estimation, with 40% retained for out-of-sample performance assessment.  

## Set-up

The research was conducted on a single machine, using an i7-9750H CPU @ 2.60GHz (12 CPUs) and 16384 MB RAM, plus NVIDIA GeForce RTX 260 GPU, with 1920 CUDA cores, a memory data rate of 14.00 Gbps, and total graphics memory of 14259 MB. I implemented certain functions in the the model with GPU acceleration, using CuPy libraries. I implemented Numba, a just-in-time (JIT) compiler, to translate some of my Python functions to optimized machine code at runtime, in turn achieved via use of LLVM compiler libraries. I also used memoization, a technique for recording intermediate results to avoid repeated calculations. This set up reduced the compute time versus a baseline Python implementation of the MATLAB code of Calvet & Fisher (2013) on the same machine from 00:08:51.624 [hh:mm:ss], or nearly 9 minutes, to 00:00:00.927 [hh:mm:ss], or just under 1 second (comparison based upon the benchmark JPYUSD daily returns dataset with $\bar k = 4$).
$\\[0.1in]$

The longest compute time requirement for this research was 162:37:06 [hh:mm:ss], to estimate EURUSD with $\bar k = 13$. Why so long? Denote $M_t$ as a volatility state vector, where each component $M_{k, t}$ is a state variable. Recall that the econometrician observes the returns $x_t = \sigma {\left[ g \left( M_t \right) \right]}^\frac {1} {2} \epsilon_t$, but not the vector $M_t$ itself; the vector $M_t$ is latent and thus inferred recursively by Bayesian updating (see section 6.3). From an implementation perspective, the Bayesian recursion is one of two major obstacles to successful implementation of MSM with HF data (the other being the construction of the transition matrix $A$). Since it is not possible to vectorize the recursion, other means of speeding up the recursion must be found (NB: Vectorized code performs operations upon multiple components of a vector at the same time (in one statement) instead of looping. Vectorizing loops can be extremely difficult, but can reduce very long processing times to virtually instantaneous). In my review of the multifractal literature I found that 13 of 31 multifractal papers used daily or weekly prices and 6 used simulated data only. 8 studies were theoretical and used no data. Only 5 studies used HF data, but none of the studies using HF data analytically derived an exact likelihood. Typically datasets used for estimating MSM and forecasting consisted of a few thousand observations. In this context, a recursion (loop) does not impose an unacceptable computational overhead. But with HF datasets approaching hundreds of thousands of observations or more, it very quickly becomes intractable, especially so when each iteration of the loop involves the fresh construction of a high-dimensional state space (i.e. a high value, say $\geq 11$, has been ascribed to $\bar k$). 
$\\[0.1in]$

I firstly discuss results for the individual parameters. In table 10 it can be seen that the multiplier $\hat m_0$ decreases with $\bar k$, in line with theory, as the amount of variability required within in each component falls in converse to the increase in the number of components. This behavior, along with the absolute values for $\hat m_o$ and the range of values for $\bar k = 4, ..., 10$ are consistent with values reported in multifractal literature for both univariate and bivariate MSM settings (see for example Calvet, Fearnley, Fisher & Leopold (2015), Alausa (2014), Idier (2008), Calvet, Fisher & Thomson (2006), and Calvet & Fisher (2004). The values of $\hat m_0$, and thus the importance of stochastic volatility, are largest for AAPL and JPM, smaller for daily FX, and smaller still for EURUSD (HF FX).  When $\bar k = 13$ for EURUSD, the value of $\hat m_0$ has decreased to 1.29904, which is the closest to 1 among the four sets of results. This indicates that the MSM captures well the long-run volatility level, which is more stable for the EURUSD, so that the regime-switching between volatility frequencies is used more to capture return outliers than is the case for AAPL and JPM. 
$\\[0.1in]$

Estimates for $\hat \sigma$ decrease, almost monotonically, for all three HF datasets. This indicates that when HF data is used, MSM is able to find a long-run average of volatility once a moderate number of components are introduced. The steps taken to mitigate microstructure noise (see section 5.1) are likely to be a factor in this observation.  The behavior of $\hat \sigma_{\mathrm{AAPL, JPM, EURUSD}}$ when microstructure noise is accounted for contrasts notably with $\hat \sigma_{\mathrm{DEXJPUS}}$, which shows much more variability, but with no particular pattern of increasing or decreasing. Empirical literature on estimation of MSM is confined to daily returns and the behavior of $\hat \sigma$ for DEXJPUS is consistent with it (for example, see Calvet, Fearnley, Fisher & Leopold (2015), Alausa (2014), Idier (2008), Calvet, Fisher & Thomson (2006), and Calvet & Fisher (2004)).
$\\[0.1in]$

As $\bar k$ increases, all HF datasets show the highest frequency multiplier, $\hat \gamma_{\bar k}$, switching more often and the spacing between frequencies becoming tighter. This is also the case for daily returns, although less pronounced, with DEXJPUS rising for $\bar k = 4, ..., 6$, then oscillating in a relatively tight range around 0.9 for moderate and high $\bar k$. With the exception of $\bar k = 4$, $\gamma_{\bar k}$ is higher for EURUSD than the HF equities data, AAPL and JPM, so that longer cycles in the FX data are predominant, with longer cycles still predominant in the daily FX returns dataset DEXJPUS. The frequency growth rate, $b$, decreases steadily with $\bar k$, with the range of values from highest (i.e. at $\bar k = 4$) to lowest ($\bar k = 13$) greatest for EURUSD. This indicates that changes occur at higher frequency and the spacing of volatility states across frequencies becomes closer. These findings are generally consistent with Calvet, Fearnley, Fisher & Leopold (2015), Alausa (2014), Idier (2008), and Calvet & Fisher (2004)), not withstanding this literature is based upon daily returns and less volatility frequencies.
$\\[0.1in]$

I now turn to the likelihood. The behavior of the log-likelihood function across the four datasets confirmed one of the main premises of the multifractal approach: namely, volatility fluctuations have a multiplicity of frequencies and explicitly incorporating a larger number of frequencies resulted in a better fit. For all four datasets the likelihood increased steadily as the number of frequencies increased, reaching a maximum at $\bar k = 13$, indicating a steady increase of fit in-sample. 
$\\[0.1in]$

The likelihood is highest for the largest dataset, EURUSD (estimated on 60% of 762,365 return observations), next highest for AAPL (60% of 449,389), and lowest for JPM, which is also the smallest of the HF datasets. This indicates that the model is well-specified and more information in the form of larger HF datasets is put to good use by it. Comparatively few multifractal researchers have used HF data and the studies that have used HF data have focused upon durations (see Zikes, Barunik & Shenai (2018), Aldrich, Heckenbach & Laughlin (2014), and Chen, Diebold & Schorfheide (2013)) or empirical study of the multifractal characteristics of the data (see Mandelbrot, Calvet & Fisher (1997c) and Calvet & Fisher (2002)). There is one exception to this, however, the study by Bacry, Khozhemyak & Muzy (2008). This research estimated the Multifractal Random Walk (MRW) of Bacry, Delour & Muzy (2001) and included the study of an HF index dataset of some 43,000 observations. Bacry, Khozhemyak & Muzy showed that GMM estimation of MRW parameters improved with the number of observations and commented "The quality of the fit (i.e. of the estimation) depends of course on the number of points in the time series".

The empirical literature has been somewhat limited, due to the constraint of computational overhead associated with analytical likelihood estimation. Calvet & Fisher's seminal 2004 paper showed that for daily FX returns, $\mathrm{MSM} \left( 10 \right)$ outperformed models with low and moderate numbers of frequencies. For $\bar k \in \{1, ..., 3\}$ the difference is significant at the 1% level, with lower significant levels obtained for larger values of $\bar k$. In a subsequent paper, Calvet, Fisher & Thompson (2006) estimated a bivariate MSM and showed that likelihood functions sharply increased with the number of frequencies. When estimating the bivariate model, Calvet, Fisher & Thompson computed an analytical bivariate likelihood for $\bar k \in \{1, ..., 5\}$ and for $\bar k \in \{6, ..., 8\}$ simulated likelihood via a particle filter, thus also showing the particle filter to be effective in extending the range of models considered tractable (i.e. $\bar k \in \{1, ..., 5\}$) within that study. Calvet, Fearnley, Fisher & Leopold (2015) extended the MSM process of Calvet & Fisher (2004) to account for the variation in skewness and term-structure of option data. With index returns, in-sample results showed that the likelihood function increased with the number of volatility components. Alausa (2014) estimated a bivariate MSM and showed $\mathrm{VEC-MSM} \left( 8 \right)$ significantly outperformed models with 1 to 7 frequencies for heating oil, corn, gold, and GBPUSD. However, results for WTI, wheat, silver, CADUSD and index data showed that lower and intermediate numbers of frequencies fitted the data equally well. Idier's (2008) results for univariate MSM also showed that lower and intermediate numbers of frequencies fit the data equally well.

Literature supports the observed strong perfomance of $\mathrm{MSM} \left( 13 \right)$. Firstly, in the first forecasting analysis of MSM, Calvet and Fisher (2004) showed that the binomial MSM matched or slightly outperformed GARCH and MS-GARCH at short horizons (1-day and 5-day), and in many cases substantially dominated these models at longer horizons (10, 20, and 50-days), using daily FX returns data. Calvet & Fisher posited that the MSM process did well because it is well described. The comparison proffered by figures 12 and 13 is quite telling in this respect, showing clearly how well MSM captures the features of financial returns series. MSM represents data well among many different dimensions. Whereas standard models require the explanation of several statistical phenomena - for example regime-switching, linear dynamics, long-memory, thick tails - MSM offers a much more concise explanation. My results suggest that increasing the state space and specifically mitigating microstructure noise provide useful additional descriptions of HF financial data within a multifractal framework.
$\\[0.1in]$

Secondly, Patton & Sheppard (2015) showed sizable and significant gains for predicting equity returns volatility by incorporating signed high frequency volatility information. Patton & Sheppard's analysis was based on realized semivariance estimators (Barndorff-Nielsen, Kinnebrock, and Shephard (2010)) and showed that relatively simple estimators allowed the decomposition of realized volatility into a part coming from positive high frequency returns and a part coming from negative high frequency returns. Disentangling the effects of these two components was shown to improve forecasts of future volatility. Accounting for imbalances via the WMP achieves a similar outcome.
$\\[0.1in]$

Thirdly, Andersen, Bollerslev & Meddahi (2011) analyzed RV forecasts that specifically took into account microstructure frictions, by building upon the eigenfunction representation of stochastic volatility of Meddahi (2001). They found that microstructure noise was substantially detrimental to forecasting accuracy; further, they also reported that linear forecasts obtained by averaging standard sparsely sampled realized volatility measures generally performed on par with more robust measures. The steps I took to cleanse and prepare HF data and thus mitigate microstructure noise, along with the 1-minute sparse sampling, may have similarly contributed to $\mathrm{MSM} \left( 13 \right)$'s strong performance.
$\\[0.1in]$

Fourthly, Lux (2008) introduced an alternative GMM estimator, thus enabling linear forecasting, and compared GMM plus linear forecasting to MLE and optimal forecasts. Lux found that GMM was generally less efficient than ML and was typically slightly disadvantaged against the ML-based forecasts. But he also showed that both GMM and linear forecasts benefited from increasing the number of volatility components. Many other studies showed this too (see for example Calvet, Fearnley, Fisher & Leopold (2015), Alausa (2014), Idier (2008), Calvet, Fisher & Thomson (2006), and Calvet & Fisher (2004), but Lux demonstrated a monotonic improvement in performance, with GMM, up to 15 volatility frequencies.
$\\[0.1in]$

# Conclusion

I conclude with a summary of my main insights and contributions. Firstly, I expanded the state space to allow more volatility components to be applied in likelihood-based estimation and forecasting. Secondly, I enabled this to be done with HF data. Thirdly, I specifically accounted for and mitigated microstructure noise. Fourthly, to enable the previous three contributions, I introduced several implementation-related developments to speed up the model. This all resulted in a well-specified model, better able to use the increased information provided to it by large amounts of HF data. Fifthly, in-sample model selection tests showed statistically significant monotonic improvement in the model as more volatility components were introduced. Sixthly, MSM(13) was compared to the relative accuracy of out-of-sample forecasts produced by a RV measure using the heterogeneous autoregressive (HAR) model of Corsi (2009). MSM(13) provided comparatively better, statistically significant, forecasts than HAR most of the time at 1-hour, 1-day, and 2-day horizons for equity HF (Apple and J.P.Morgan) and foreign exchange HF (EURUSD) returns series. MZ regressions showed little sign of bias in the MSM(13) forecasts.
$\\[0.1in]$

These insights and contributions suggest MSM may provide a viable alternative to established realized volatility estimators in high-frequency settings, which opens up many potential avenues for further research. They also provide a natural bridge to non-linear state space methods in the machine learning canon, which I investigate subsequently in a companion paper.
$\\[0.1in]$
